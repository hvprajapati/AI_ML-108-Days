# Python APIs Masterclass
## Fetch Live Data from the Internet

An API (Application Programming Interface) lets your Python code talk to other services -- fetch weather, stock prices, news, or any web data. This notebook uses the free `requests` library.

## 1. Setup

```python
# pip install requests
import requests
print("Requests ready!")
```

## 2. Your First API Call -- HTTP GET

```python
# Fetch data from a free public API
response = requests.get("https://api.github.com")

print("Status Code:", response.status_code)  # 200 = OK, 404 = Not Found
print("Content Type:", response.headers['content-type'])
print("Response (first 300 chars):")
print(response.text[:300])
```

### HTTP Status Codes
- 200 = Success (OK)
- 201 = Created
- 400 = Bad Request (your fault)
- 404 = Not Found
- 500 = Server Error (their fault)

## 3. Working with JSON APIs

```python
# Most APIs return JSON
response = requests.get("https://api.github.com/users/python")
data = response.json()  # Parse JSON automatically!

print("Name:", data.get('name'))
print("Followers:", data.get('followers'))
print("Public Repos:", data.get('public_repos'))

# The response is just a Python dict -- use it like any dict
for key in ['name', 'company', 'blog', 'location']:
    print(f"  {key}: {data.get(key, 'N/A')}")
```

## 4. Handling Errors Gracefully

```python
def fetch_json(url):
    """Safely fetch JSON from a URL."""
    try:
        response = requests.get(url, timeout=10)
        response.raise_for_status()  # Raises error for 4xx/5xx
        return response.json()
    except requests.exceptions.ConnectionError:
        print("Error: No internet connection!")
    except requests.exceptions.Timeout:
        print("Error: Request timed out!")
    except requests.exceptions.HTTPError as e:
        print(f"Error: {e}")
    return None

# Test with a bad URL
fetch_json("https://api.github.com/users/this-does-not-exist-999")
```

## 5. Free Public APIs to Practice With

These APIs need no key -- just a URL:

| API | URL | Returns |
|-----|-----|--------|
| Random Cat Fact | https://catfact.ninja/fact | Cat fact |
| Random User | https://randomuser.me/api/ | Fake profile |
| REST Countries | https://restcountries.com/v3.1/name/india | Country info |
| Dog Images | https://dog.ceo/api/breeds/image/random | Dog pic URL |

```python
# Example: Get a random cat fact
resp = requests.get("https://catfact.ninja/fact")
if resp.status_code == 200:
    print("Cat Fact:", resp.json()['fact'])

# Example: Get a fake user from JSONPlaceholder (always works, no key needed)
resp = requests.get("https://jsonplaceholder.typicode.com/users/1")

if resp.status_code == 200:
    user = resp.json()         
    print(f"Name:    {user['name']}")
    print(f"Email:   {user['email']}")
    print(f"City:    {user['address']['city']}")
    print(f"Company: {user['company']['name']}")
```

## 6. Mini Project -- Random User Generator

```python
def get_random_user():
    """Fetch and display a random user."""
    resp = requests.get("https://randomuser.me/api/")
    if resp.status_code == 200:
        user = resp.json()['results'][0]
        name = f"{user['name']['first']} {user['name']['last']}"
        email = user['email']
        country = user['location']['country']
        return f"{name} | {email} | {country}"
    return "Failed to fetch user"

print(get_random_user())
print(get_random_user())
```

# Summary

- `requests.get(url)` -- fetch data from an API
- `.json()` -- parse JSON response to Python dict
- `.status_code` -- check if request succeeded (200 = OK)
- `.raise_for_status()` -- auto-check for errors
- Always wrap API calls in `try/except` for network errors

APIs connect your Python code to the entire internet. This is how real applications work!
