# PART 1: JSON (JavaScript Object Notation)

## 1. What is JSON?

JSON is a lightweight text format for storing and exchanging data. It looks like Python dictionaries and lists.

```json
{
  "name": "Rahul",
  "age": 22,
  "skills": ["Python", "ML"]
}
```

**Why JSON?** Every API, web service, and config file uses it. It is the universal language of data exchange.

## 2. Python dict/list -> JSON String (Serialization)

### `json.dumps(obj)`
Converts a **Python object (dict, list, etc.)** into a **JSON string**.

> **Easy trick:** `dumps` = **Dict → JSON String**

````python
# Import JSON module for reading/writing JSON data
import json

# Python dict
student = {
    "name": "Rahul",
    "age": 22,
    "city": "Ahmedabad",
    "marks": [85, 90, 78]
}

# Convert to JSON string
# Convert Python dict/list to JSON string
json_string = json.dumps(student)
print(json_string)
print(type(json_string))
````

````python
# Pretty-print with indentation
print(json.dumps(student, indent=4))
````

## 3. JSON String -> Python dict/list (Deserialization)

````python
json_data = '{"name": "Priya", "age": 20, "city": "Mumbai"}'

# Parse JSON string to Python dict
# Parse JSON string into Python dict/list
person = json.loads(json_data)
print(person["name"])
print(person["city"])
print(type(person))
````

## 4. Reading & Writing JSON Files

````python
# WRITE: Save Python data to a JSON file
library = {
    "name": "City Library",
    "books": [
        {"id": 1, "title": "Python Basics", "author": "Guido"},
        {"id": 2, "title": "Data Science 101", "author": "Jane"}
    ]
}

# Open file (automatically closes when done)
with open("library.json", "w") as f:
# Write Python object to a JSON file
    json.dump(library, f, indent=4)

print("Saved to library.json!")
````

````python
# READ: Load JSON file back into Python
# Open file (automatically closes when done)
with open("library.json", "r") as f:
# Read JSON data from a file into Python
    loaded_data = json.load(f)

print(loaded_data["name"])
for book in loaded_data["books"]:
    print(f"  - {book['title']} by {book['author']}")
````

## 5. Real-World JSON -- Working with an API Response

````python
# This is what a typical API response looks like
api_response = '''
{
  "status": "success",
  "data": {
    "users": [
      {"id": 1, "name": "Rahul", "active": true},
      {"id": 2, "name": "Priya", "active": false},
      {"id": 3, "name": "Amit", "active": true}
    ]
  }
}
'''

# Parse JSON string into Python dict/list
response = json.loads(api_response)

# Extract only active users using a comprehension
active_users = [u["name"] for u in response["data"]["users"] if u["active"]]
print("Active users:", active_users)
````

# PART 2: CSV (Comma Separated Values)

## 6. What is CSV?

CSV is the simplest data format -- a table stored as plain text. Excel, Google Sheets, and databases all support CSV.

```
name,age,city
Rahul,22,Ahmedabad
Priya,20,Mumbai
```

## 7. Reading CSV Files

````python
# Import CSV module for reading/writing CSV files
import csv

# First, create a sample CSV file
# Open file (automatically closes when done)
with open("students.csv", "w", newline="") as f:
# Create a CSV writer object (writes lists)
    writer = csv.writer(f)
# Write a single row to CSV file
    writer.writerow(["name", "age", "marks"])
# Write a single row to CSV file
    writer.writerow(["Rahul", 22, 85])
# Write a single row to CSV file
    writer.writerow(["Priya", 20, 92])
# Write a single row to CSV file
    writer.writerow(["Amit", 21, 78])

print("Created students.csv")
````

````python
# Now read it back
# Open file (automatically closes when done)
with open("students.csv", "r") as f:
# Create a CSV reader object (returns lists)
    reader = csv.reader(f)
    for row in reader:
        print(row)
````

````python
# Better: Read as list of dictionaries
# Open file (automatically closes when done)
with open("students.csv", "r") as f:
# Read CSV rows as dictionaries (column names as keys)
    reader = csv.DictReader(f)
    for row in reader:
        print(f"{row['name']} scored {row['marks']} marks")
````

## 8. Writing CSV Files

````python
# Write using a list of dictionaries
data = [
    {"product": "Laptop", "price": 50000, "stock": 10},
    {"product": "Mouse", "price": 800, "stock": 50},
    {"product": "Keyboard", "price": 1500, "stock": 30}
]

# Open file (automatically closes when done)
with open("inventory.csv", "w", newline="") as f:
# Write dictionaries to CSV (column names from fieldnames)
    writer = csv.DictWriter(f, fieldnames=["product", "price", "stock"])
# Write column headers to CSV file
    writer.writeheader()
# Write multiple rows to CSV file at once
    writer.writerows(data)

print("Created inventory.csv")

# Verify
# Open file (automatically closes when done)
with open("inventory.csv", "r") as f:
    print(f.read())
````

## 9. JSON vs CSV -- When to Use Which

| Feature | JSON | CSV |
|---------|------|-----|
| Structure | Nested (trees, lists) | Flat (table) |
| Best for | APIs, configs, nested data | Spreadsheets, tabular data |
| Python | json.loads() / dumps() | csv.reader() / writer() |
| File size | Larger | Smaller |

## 10. Error Handling with JSON

````python
# Bad JSON will crash your program -- handle it!
bad_json = '{"name": "Rahul", age: 22}'  # Missing quotes around age

try:
# Parse JSON string into Python dict/list
    data = json.loads(bad_json)
# Handle broken/malformed JSON gracefully
except json.JSONDecodeError as e:
    print(f"Invalid JSON: {e}")
````
