## Function Execution Flow

```
Call Function
      │
      ▼
Execute statements
      │
      ▼
Return to caller
```

## Parameters vs Arguments

- **Parameter**: variable in function definition.
- **Argument**: actual value passed.

````python
def greet(name):
    print(f"Hello, {name}")

greet("Alice")
greet("Bob")
````

## Positional vs Keyword Arguments

````python
def introduce(name, age):
    print(name, age)

introduce("John", 25)
introduce(age=30, name="Hardik")
introduce("Ronak", age=60)
````

## Default Parameters

````python
def power(base, exponent=2):
    return base ** exponent

print(power(5))
print(power(5,3))
````

## Return Statement

````python
def add(a,b):
    return a + b

add(a=5, b=10)
````

## print() vs return()

````python
def using_print():
    print(5)

def using_return():
    return 5

x = using_print()
y = using_return()

print(x)
print(y)
````

## Returning Multiple Values

````python
def stats(a,b):
    return a+b, a*b

s,p = stats(5,6)
print(s,p)
````

## Local Variables

````python
def demo():
    x=10
    print(x)

demo()
````

## Global Variables

````python
count=0

def increment():
    global count
    count+=1

increment()
print(count)
````

## Type Hints

````python
def multiply(a:int,b:int)->int:
    return a*b

print(multiply(3,4))
````

## Common Mistakes

- Forgetting to call the function
- Mixing parameter and argument
- Using `print()` instead of `return`
- Returning too early
- Changing global variables accidentally

## Mini Project - Student Grade Calculator

````python
def calculate_grade(marks):
    avg = sum(marks)/len(marks)

    if avg>=90:
        grade='A'
    elif avg>=75:
        grade='B'
    elif avg>=60:
        grade='C'
    else:
        grade='D'
    return avg, grade

marks=[85,90,78,88]
avg,grade=calculate_grade(marks)

print("Average:",avg)
print("Grade:",grade)
````
