## While Loops, Loop Control Statements & Advanced Looping

# 1. Why Do We Need a `while` Loop?

A `for` loop is great when you know **how many times** to repeat.

A `while` loop is useful when you **don't know in advance** how many times the loop should run.

Examples:
- ATM menu
- Login system
- Guess the number game
- Reading data until the user quits


## 2. Basic `while` Loop

```python
count = 1

while count <= 5:
    print(count)
    count += 1
```

## 3. Infinite Loops

```python
# Be careful! This would run forever.
# while True:
#     print("Running...")

# Correct approach:
count = 1
while count <= 3:
    print(count)
    count += 1
```

## 4. `break` Statement

```python
for number in range(1, 11):
    if number == 6:
        break
    print(number)
```

```python
password = "python"

while True:
    user = input("Enter password: ")
    if user == password:
        print("Access Granted!")
        break
    print("Try Again")
```

## 5. `continue` Statement

```python
for number in range(1,11):
    if number % 2 == 0:
        continue
    print(number)
```

## 6. `pass` Statement

```python
for i in range(3):
    if i == 1:
        pass
    print(i)
```

## 7. `else` with Loops  :
## In Python, an else block attached to a loop executes only if the loop completes all its iterations normally.

```python
for i in range(5):
    print(i)
else:
    print("Loop completed successfully!")
```

```python
for i in range(5):
    if i == 3:
        break
    print(i)
else:
    print("This never executes because break was used.")
```

## 8. `while ... else`

```python
count = 1
while count <= 3:
    print(count)
    count += 1
else:
    print("While loop finished.")
```

## 9. Advanced Looping Techniques

```python
students = ["Alice","Bob","Charlie"]

for index, name in enumerate(students, start=1):
    print(index, name)
```

```python
names = ["Alice","Bob","Charlie"]
marks = [90,85,95]

for name, mark in zip(names, marks):
    print(name, mark)
```

```python
students = {
    "Alice":90,
    "Bob":85,
    "Charlie":95
}

for name, mark in students.items():
    print(name, mark)
```

## 10. Pattern Printing

```python
for i in range(5,0,-1):
    print("*"*i)
```

```python
rows = 5
for i in range(rows):
    spaces = " " * (rows-i-1)
    stars = "*" * (2*i+1)
    print(spaces + stars)
```

```python
for i in range(1,6):
    print(str(i)*i)
```

## 11. Mini Project - Guess the Number

```python
secret = 7

while True:
    guess = int(input("Guess the number: "))

    if guess == secret:
        print("Correct!")
        break
    elif guess < secret:
        print("Too small")
    else:
        print("Too large")
```

## 12. Mini Project - ATM Menu

```python
balance = 1000

while True:
    print("\n1. Check Balance")
    print("2. Deposit")
    print("3. Exit")

    choice = input("Choice: ")

    if choice == "1":
        print("Balance:", balance)
    elif choice == "2":
        amount = int(input("Amount: "))
        balance += amount
    elif choice == "3":
        print("Thank you!")
        break
    else:
        print("Invalid Choice")
```

## 13. Common Mistakes

- Forgetting to update the loop variable.
- Creating accidental infinite loops.
- Using `continue` before updating variables in a `while` loop.
- Misunderstanding `break` vs `continue`.
- Forgetting that `else` won't execute if `break` is used.
