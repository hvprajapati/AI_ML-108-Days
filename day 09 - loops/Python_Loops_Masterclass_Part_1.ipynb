## For Loops from Absolute Beginner to Intermediate

This notebook teaches **why loops exist**, how `for` loops work, `range()`, nested loops, and plenty of examples.

# 1. Why Do We Need Loops?

Imagine you want to print `"Hello"` 100 times.

Without loops:

```python
print("Hello")
print("Hello")
...
```

This is repetitive and hard to maintain.

A **loop** allows us to repeat code automatically.


# 2. What is a Loop?

A loop repeatedly executes a block of code until it finishes iterating over data or a condition becomes false.

Execution Flow:

```
Start
  │
  ▼
Loop Starts
  │
  ▼
Execute Code
  │
  ▼
Next Item?
 │
Yes ───────────┐
 │             │
 └─────────────┘
No
 │
 ▼
End
```


# 3. Your First `for` Loop

````python
for i in range(5):
    print("Hello World")
````

### Explanation
`range(5)` produces numbers from 0 to 4. The loop runs once for each number.

# 4. Loop Variable

````python
for number in range(5):
    print(number)
````

# 5. Iterating Over Different Collections

````python
fruits = ["Apple", "Banana", "Mango"]
for fruit in fruits:
    print(fruit)
````

````python
name = "Hardik"

for letter in name:
    print(letter)
````

````python
student = {
    "name":"Hardik",
    "age":22
}

for key in student:
    print(key, "->", student[key])
````

````python
colors = {"Red","Green","Blue"}

for color in colors:
    print(color)
````

# 6. Understanding `range()`

````python
print(list(range(5)))
print(list(range(2,8)))
print(list(range(2,20,3)))
print(list(range(10,0,-2)))
````

`range(stop)`

`range(start, stop)`

`range(start, stop, step)`

# 7. Nested Loops

````python
for i in range(3):
    for j in range(3):
        j=i
        print(f"i={i}, j={j}")
````

# 8. Pattern Printing

````python
for i in range(1,6):
    print("*"*i)
````

````python
for i in range(1,6):
    for j in range(i):
        print(j+1,end="")
    print()
````

# 9. Useful Real-World Examples

````python
marks=[80,95,72,88]

total=0

for mark in marks:
    total+=mark

print("Average:", total/len(marks))
````

````python
numbers=[5,7,10,14,20]

largest=numbers[0]

for num in numbers:
    if num>largest:
        largest=num

print(largest)
````
