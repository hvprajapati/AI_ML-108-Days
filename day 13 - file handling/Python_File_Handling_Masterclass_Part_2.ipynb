## 2. Working with CSV Files

```python
import csv

rows = [
    ["Name","Marks"],
    ["Alice",90],
    ["Bob",85]
]

with open("students.csv","w",newline="",encoding="utf-8") as f:
    # Python will create this file if it doesn't already exist.
    writer = csv.writer(f)
    writer.writerows(rows)

print("CSV created.")
```

```python
import csv

with open("students.csv","r",encoding="utf-8") as f:
    reader = csv.reader(f)
    for row in reader:
        print(row)
```

## 3. CSV using DictReader / DictWriter

```python
import csv

with open("employees.csv","w",newline="",encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=["name","salary"])
    writer.writeheader()
    writer.writerow({"name":"Hardik","salary":50000})

with open("employees.csv","r",encoding="utf-8") as f:
    for row in csv.DictReader(f):
        print(row)

```

## 4. Working with JSON

```python
import json

student = {
    "name":"Hardik",
    "age":22,
    "skills":["Python","SQL","AWS"]
}

with open("student.json","w",encoding="utf-8") as f:
    json.dump(student,f,indent=4)
```

```python
import json

with open("student.json","r",encoding="utf-8") as f:
    data=json.load(f)

print(data)
print(data["skills"])
```

## 5. Binary Files

```python
data=b"Hello Binary World"

with open("sample.bin","wb") as f:
    f.write(data)

with open("sample.bin","rb") as f:
    print(f.read())

print(type(data))
# print(list(data))
```

## 6. Serialization with Pickle

pickle is used to save Python objects to a file and load them back later

```python
import pickle

student={"name":"Hardik","marks":[90,95,88]}

with open("student.pkl","wb") as f:
    pickle.dump(student,f)

with open("student.pkl","rb") as f:
    restored=pickle.load(f)

print(restored)
```

## 7. Working with os

```python
import os

print(os.getcwd())
print(os.path.exists("student.json"))
print(os.listdir("."))
```

## 8. Using pathlib

```python
from pathlib import Path

path=Path("student.json")

print(path.exists())
print(path.name)
print(path.suffix)
print(path.absolute())
```

## 9. Copying Files with shutil

```python
import shutil

shutil.copy("student.json","student_backup.json")
print("Backup created.")
```

## 10. Processing Large Files Efficiently

```python
with open("notes.txt","r",encoding="utf-8") as f:
    for line in f:
        print(line.strip())
```

Reading line-by-line is memory efficient compared to `read()` for very large files.

## 12. Mini Project - CSV Marks Analyzer

```python
import csv

total=0
count=0

with open("students.csv","r",encoding="utf-8") as f:
    reader=csv.reader(f)
    next(reader)
    for _,marks in reader:
        total+=int(marks)
        count+=1

print("Average:",total/count)
```

## 13. Mini Project - JSON Configuration Loader

```python
import json

config={
    "host":"localhost",
    "port":8000
}

with open("config.json","w") as f:
    json.dump(config,f)

with open("config.json") as f:
    cfg=json.load(f)

print(cfg)
```

## 14. Best Practices


- Always use `with`.
- Specify `encoding="utf-8"` for text files.
- Handle `FileNotFoundError`.
- Prefer `pathlib` for new projects.
- Stream large files instead of loading them entirely.
- Never unpickle untrusted files.


## 15. Common Mistakes


- Forgetting `newline=""` when writing CSV files.
- Mixing text and binary modes.
- Assuming JSON supports every Python object.
- Using pickle with untrusted data.
- Hardcoding absolute paths.
