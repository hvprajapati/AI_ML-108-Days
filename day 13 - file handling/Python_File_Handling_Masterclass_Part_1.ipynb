# 1. Why Do We Need File Handling?

Variables store data only while the program is running.

If you want data to remain after the program exits, you need **files**.

Examples:
- Student records
- Configuration files
- Logs
- Reports
- CSV datasets


# 2. What is a File?

A **file** is a collection of data stored on disk.

Two common types:
- **Text files** (`.txt`, `.csv`, `.json`, `.py`)
- **Binary files** (`.jpg`, `.png`, `.pdf`, `.exe`)


## 3. Opening a File

````python
file = open("sample.txt", "w",encoding="utf-8")
file.write("Hello File Handling!")
file.close()
````

### Syntax

```python
open(filename, mode, encoding="utf-8")
```


## 4. File Modes


| Mode | Meaning |
|------|---------|
| `r` | Read |
| `w` | Write (overwrite) |
| `a` | Append |
| `x` | Create new file |
| `r+` | Read & Write |
| `w+` | Write & Read |
| `a+` | Append & Read |
| `rb` | Read Binary |
| `wb` | Write Binary |
| `ab` | Append Binary |


## 5. Writing to a File

````python
with open("notes.txt", "w", encoding="utf-8") as file:
    file.write("Python File Handling\n")
    file.write("Learning is fun!")  
````

## 6. Reading Files

````python
with open("notes.txt", "r", encoding="utf-8") as file:
    print(file.read())
````

## 7. Reading Limited Characters

````python
with open("notes.txt", "r", encoding="utf-8") as file:
    print(file.read(6))
````

## 8. Reading One Line

````python
with open("notes.txt", "r", encoding="utf-8") as file:
    print(file.readline())
````

## 9. Reading All Lines

````python
with open("notes.txt", "r", encoding="utf-8") as file:
    lines = file.readlines()

print(lines)
````

## 10. Reading Using a Loop

````python
with open("notes.txt", "r", encoding="utf-8") as file:
    for line in file:
        print(line.strip())
````

## 11. Appending Data

````python
with open("notes.txt", "a", encoding="utf-8") as file:
    file.write("\nThis line was appended.")
````

## 12. File Cursor

````python
with open("notes.txt", "r", encoding="utf-8") as file:
    print(file.tell())
    print(file.read(6))
    print(file.tell())
    file.seek(0)
    # # seek() ->  to move a pointer to the specific position
    print(file.read(6))
````

## 14. Why Use `with`?

````python
# Recommended
with open("notes.txt", "r", encoding="utf-8") as file:
    print(file.read())

# File closes automatically after leaving the block.
````


Benefits:
- Automatically closes files
- Cleaner code
- Safer when exceptions occur


## 15. Common Exceptions

````python
try:
    with open("missing.txt", "r") as file:
        print(file.read())
except FileNotFoundError:
    print("File not found.")
except PermissionError:
    print("Permission denied.")
````

## 16. Mini Project - Word Counter

````python
with open("notes.txt", "r", encoding="utf-8") as file:
    text = file.read()

words = text.split()

print("Characters:", len(text))
print("Words:", len(words))
print("Lines:", text.count("\n") + 1)
````

## 17. Mini Project - Student Record Writer

````python
students = [
    "Alice,90",
    "Bob,82",
    "Charlie,95"
]

with open("students.txt", "w", encoding="utf-8") as file:
    for student in students:
        file.write(student + "\n")

print("Student records saved.")
````
