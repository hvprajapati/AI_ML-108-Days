# Introduction to Lists

## What is a List?
A list is an ordered, mutable collection.

```python
numbers=[10,20,30]
print(numbers)
```

# Creating Lists

## Creating Lists

```python
a=[]
b=list()
c=[1,2,3]
print(a,b,c)
```

# Indexing

## Positive and Negative Indexing

```python
lst=[10,20,30,40]
print(lst[0])
print(lst[-1])
```

# Slicing

## List Slicing

```python
lst=[10,20,30,40,50]
print(lst[1:4])
print(lst[::-1])
```

# Mutability

## Lists are Mutable

```python
lst=[1,2,3]
lst[1]=999
print(lst)
```

# append

## append()

```python
lst=[1,2,3]
lst.append(4)
print(lst)
```

# extend

## extend()

```python
lst=[1,2]
lst.extend([3,4])
print(lst)
```

# insert

## insert()

```python
lst=[1,2,3]
lst.insert(1,100)
print(lst)
```

# remove

## remove()

```python
lst=[1,2,3,2]
lst.remove(2)
print(lst)
```

# pop

## pop()

```python
lst=[1,2,3]
print(lst.pop())
print(lst)
```

# clear

## clear()

```python
lst=[1,2,3]
lst.clear()
print(lst)
```

# index

## index()

```python
lst=[10,20,30]
print(lst.index(20))
```

# count

## count()

```python
lst=[1,1,1,2]
print(lst.count(1))
```

# sort

## sort()

```python
lst=[5,2,8,1]
lst.sort()
print(lst)
```

# sorted

## sorted()

```python
lst=[5,2,8,1]
print(sorted(lst))
```

# reverse

## reverse()

```python
lst=[1,2,3]
lst.reverse()
print(lst)
```

# Copying Lists

## copy() vs Reference

```python
a=[1,2,3]
b=a
c=a.copy()
print(a,b,c)
```

# Nested Lists

## Nested Lists

```python
matrix=[[1,2],[3,4]]
print(matrix[1][0])
```

# List Comprehension

## List Comprehension

```python
squares=[x*x for x in range(10)]
print(squares)
```

# Conditional Comprehension

## Conditional List Comprehension

```python
evens=[x for x in range(20) if x%2==0]
print(evens)
```

# Frequency Counter

## Frequency Counter

```python
lst=[1,2,2,3,3,3]
freq={}
for i in lst:
    freq[i]=freq.get(i,0)+1
print(freq)
```

# Second Largest

## Second Largest Element

```python
lst=[10,40,20,50]
print(sorted(lst)[-2])
```

# Remove Duplicates

## Remove Duplicates

```python
lst=[1,1,2,2,3]
print(list(set(lst)))
```
