## Recursion

A recursive function is a function that calls itself. Every recursive solution needs:
1. Base case (stop condition)
2. Recursive case (smaller problem)

````python
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n-1)

print(factorial(5))
````

````python
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)

for i in range(8):
    print(fibonacci(i), end=" ")
````

## Decorators

A decorator lets you add functionality to an existing function without changing its source code.
A decorator is a function that takes another function as an argument and returns a new function that modifies the behavior of the original function. The new function is often referred to as a "decorated" function.

````python
def logger(func):
    def wrapper():
        print("Before function")
        func()
        print("After function")
    return wrapper

@logger
def greet():
    print("Hello!")

greet()
````

````python
logger(greet)()
````

## Generators

Generators produce values one at a time using `yield`.

````python
def numbers():
    print("Start")
    yield 1

    print("Middle")
    yield 2

    print("End")
    yield 3
````

````python
nums = numbers()

print(next(nums))
print(next(nums))

print(next(nums))

````

````python
def numbers():
    yield 1
    yield 2
    yield 3

for num in numbers():
    print(num)

# Python automatically calls next() behind the scenes.
````

````python
squares = (x*x for x in range(6))
for i in squares:
    print(i)
````

## Iterators vs Generators

Every generator is an iterator, but not every iterator is a generator.

````python
nums=[1,2,3]
it=iter(nums)

print(next(it))
print(next(it))
print(next(it))
````

````python
numbers = [10, 20, 30]

for num in numbers:
    print(num)


# looks simple right
# But Python secretly does this:
````

````python
numbers = [10, 20, 30]

it = iter(numbers)

while True:
    try:
        num = next(it)
        print(num)
    except StopIteration:
        break
````

## Function Factory

A function that creates and returns another function.

````python
def power(exponent):
    def calculate(base):
        return base ** exponent
    return calculate

square = power(2)
cube = power(3)

print(square(8))
print(cube(4))
````

## Mini Project - Expense Tracker

````python
def total_expense(expenses):
    return sum(expenses)

def average_expense(expenses):
    return sum(expenses)/len(expenses)

expenses=[250,400,150,800,120]

print("Total:", total_expense(expenses))
print("Average:", average_expense(expenses))
````

````python
# Python provides memoization using the @lru_cache decorator.
from functools import lru_cache

@lru_cache
def square(n):
    print("Calculating...")
    return n * n


print(square(5))
print(square(5))
print(square(5))
````

#  Real-World Use Cases of Generators

Generators are widely used when working with **large datasets** or **continuous streams of data**. Instead of loading everything into memory at once, they generate values **one at a time**, making programs more memory-efficient and scalable.

---

## 1. Reading Large Files

When working with very large files (e.g., 10 GB log files), loading the entire file into memory is inefficient.

Instead, Python reads **one line at a time**.

```python
with open("data.txt", "r") as file:
    for line in file:
        print(line)
```

###  Why use a generator?

- Only one line is stored in memory at any given time.
- Ideal for processing large log files, CSVs, and text files.
- Prevents memory overflow.

---

## 2. Processing Database Records

Imagine a database table containing **1 million rows**.

Loading every row into memory is unnecessary if you're processing them sequentially.

Instead, fetch one record at a time.

```python
for row in database_cursor:
    process(row)
```

###  Why use a generator?

- Low memory usage.
- Faster processing for large datasets.
- Common in ETL pipelines and backend applications.

---

## 3. Paginated API Responses

Many APIs do not return all records in a single response.

Instead, they return data **page by page**.

```
Page 1
   ↓
Page 2
   ↓
Page 3
```

A generator can fetch the next page only when needed.

```python
for user in get_all_users():
    print(user)
```

###  Why use a generator?

- Handles large API responses efficiently.
- Reduces memory consumption.
- Avoids unnecessary network requests.

---

## 4. Machine Learning & Deep Learning

Datasets used in Machine Learning are often too large to fit into RAM.

Instead of loading every image at once, data is loaded **batch by batch**.

```
Batch 1
   ↓
Batch 2
   ↓
Batch 3
```

Frameworks such as **TensorFlow** and **PyTorch** use generators (or generator-like data loaders) extensively.

###  Why use a generator?

- Enables training on massive datasets.
- Reduces memory usage.
- Improves training efficiency.

---

## 5. Data Engineering (Most Important)

Suppose an S3 bucket contains a **500 GB CSV file**.

 Avoid loading the entire file into memory.

```python
import pandas as pd

df = pd.read_csv("sales.csv")
```

Instead, process the file in **small chunks**.

```python
import pandas as pd

for chunk in pd.read_csv("sales.csv", chunksize=10000):
    process(chunk)
```

###  Why use a generator?

- Only one chunk is loaded into memory at a time.
- Can process files much larger than available RAM.
- Common practice in ETL pipelines, AWS Glue, Spark preprocessing, and data engineering workflows.

---

#  Summary

| Use Case | Why Generators Are Useful |
|----------|---------------------------|
| Reading Large Files | Reads one line at a time instead of the entire file. |
| Database Records | Fetches one row at a time instead of loading all rows. |
| APIs | Retrieves data page by page on demand. |
| Machine Learning | Loads data batch by batch during training. |
| Data Engineering | Processes huge files in chunks instead of loading them entirely. |

---

#  Key Takeaway

A generator is useful whenever **the complete dataset is too large to fit into memory** or **you only need values one at a time**.

**Think of a generator as a tap, not a bucket.**

- **Bucket (List):** Gives you all the data at once.
- **Tap (Generator):** Gives you data only when you ask for it.
