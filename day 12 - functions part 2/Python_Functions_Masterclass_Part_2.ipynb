## *args

Accept any number of positional arguments.

```python
def total(*numbers):
    return sum(numbers)

print(total(1,2,3))
print(total(10,20,30,40))
```

## **kwargs

Accept any number of keyword arguments.

```python
def profile(**info):
    for k,v in info.items():
        print(k,":",v)

profile(name="Hardik", age=22, city="Ahmedabad")
```

## Packing & Unpacking

Collect and unpack values.

```python
nums=[1,2,3]
print(*nums)

def add(a,b,c):
    return a+b+c

print(add(*nums))
```

## Positional-only & Keyword-only Parameters

Modern function signatures.

```python
def demo(a, /, b, *, c):
    print(a,b,c)

demo(1,2,c=3)
```

## Functions as Arguments

Pass functions to other functions.

```python
def square(x):
    return x*x

def apply(fn,value):
    return fn(value)

print(apply(square,5))
```

## Nested Functions

Functions inside functions.

```python
def outer():
    def inner():
        print("Inner Function")
    inner()

outer()
```

## LEGB Rule
L = Local 
E = Enclosing
G = Global
B = Built-in

How Python resolves variable names.

```python
x="Global"
def outer():
    x="Enclosing"
    def inner():
        x="Local"
        print(x)
    inner()
outer()
print(x)
```

```python
x=10
def outer():
    x=25
    def inner():
        x=50
        print(x)
    inner()
    print(x)

print(x)

outer()
```

## Lambda Functions

Anonymous one-line functions.

```python
square=lambda x:x*x
print(square(9))
```

## map()

Apply a function to every item.

```python
nums=[1,2,3,4]
print(list(map(lambda x:x*x,nums)))
```

## filter()

Keep only matching items.

```python
nums=[1,2,3,4,5,6]
print(list(filter(lambda x:x%2==0,nums)))
```

## reduce()

Reduce a sequence to one value.

```python
from functools import reduce
nums=[1,2,3,4]
print(reduce(lambda a,b:a+b,nums))
```

## zip()

Combine iterables.

```python
names=["A","B","C"]
marks=[90,80,70]
print(list(zip(names,marks)))
```

## enumerate()

Loop with index.

```python
for i,v in enumerate(["apple","banana","mango"],start=1):
    print(i,v)
```

## callable()

Check if an object is callable.

```python
print(callable(len))
print(callable(100))
```

## Mini Project

Simple text processor.

```python
def text_stats(text):
    words=text.split()
    return {
        "Characters":len(text),
        "Words":len(words),
        "Upper":text.upper(),
        "Lower":text.lower()
    }

print(text_stats("Python functions are awesome"))
```
