# Python Matplotlib Masterclass
## Your First Steps into Data Visualization

Matplotlib is Python's most popular plotting library. If you can use Excel charts, you can learn Matplotlib. This notebook covers the 4 chart types you will actually use.

## 1. Setup & Your First Plot

### `plt.plot(x, y)`
Creates a **line graph** using the `x` values (horizontal axis) and `y` values (vertical axis).

### `plt.show()`
Displays the graph on the screen.

```python
# If not installed: pip install matplotlib
import matplotlib.pyplot as plt

# Simplest plot possible
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.plot(x, y)
plt.show()
```

## 2. Line Plot -- The Most Common Chart

Shows trends over time.

### Common `plt` Functions

- `plt.plot(x, y)` → Creates a line graph.
- `marker='o'` → Shows a marker at each data point.
- `color='blue'` → Sets the line color.
- `linewidth=2` → Sets the thickness of the line.
- `linestyle='--'` → Makes the line dashed.
- `plt.title()` → Adds a title to the graph.
- `plt.xlabel()` → Labels the X-axis.
- `plt.ylabel()` → Labels the Y-axis.
- `plt.grid(True)` → Displays grid lines.
- `plt.show()` → Displays the graph.

```python
months = [1, 2, 3, 4, 5, 6]
sales = [12000, 15000, 13500, 18000, 22000, 25000]

plt.plot(months, sales, marker='o', color='blue', linewidth=2, linestyle='--')
plt.title('Monthly Sales Report')
plt.xlabel('Month')
plt.ylabel('Sales (Rs.)')
plt.grid(True)
plt.show()
```

## 3. Bar Chart -- Comparing Categories

```python
subjects = ['Math', 'Physics', 'Chemistry', 'English', 'CS']
marks = [85, 78, 92, 88, 95]

plt.bar(subjects, marks, color=['red', 'blue', 'green', 'orange', 'purple'])
plt.title('Student Marks by Subject')
plt.xlabel('Subject')
plt.ylabel('Marks')
plt.show()

```

## 4. Pie Chart -- Parts of a Whole

### Common `plt.pie()` Parameters

- `usage` → Determines the size of each pie slice.
- `labels=languages` → Displays labels for each slice.
- `autopct='%1.1f%%'` → Shows the percentage on each slice with 1 decimal place.
- `startangle=90` → Starts the pie chart from the top (90°).
- `plt.title()` → Adds a title to the chart.
- `plt.show()` → Displays the chart.

```python
languages = ['Python', 'JavaScript', 'Java', 'C++', 'Other']
usage = [35, 25, 20, 10, 10]

plt.pie(usage, labels=languages, autopct='%1.1f%%', startangle=90)
plt.title('Programming Language Popularity')
plt.show()
```

## 5. Scatter Plot -- Relationship Between Variables

### `s` in `plt.scatter()`
`s` sets the **size of each point (marker)** in the scatter plot. Larger values create bigger points.

```python
hours_studied = [1, 2, 3, 4, 5, 6, 7, 8]
exam_scores = [45, 55, 60, 65, 70, 75, 82, 90]

plt.scatter(hours_studied, exam_scores, color='darkgreen', s=100)
plt.title('Study Hours vs Exam Score')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.show()
print("More study -> higher scores. This is called correlation.")
```

## 6. Mini Project -- Student Performance Dashboard

```python
# Create a 2x2 dashboard of charts
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
fig.suptitle('Student Performance Dashboard', fontsize=14)

# 2, 2 → 2 rows and 2 columns
# Total = 4 graph spaces
# fig → The whole figure (dashboard)
# axes → A 2×2 array containing the 4 graph areas
# figsize=(10,8) → Figure width = 10 inches, height = 8 inches



# Chart 1: Line - Progress over time
axes[0, 0].plot([1,2,3,4,5,6], [60,65,70,75,80,85], marker='o')
axes[0, 0].set_title('Progress Over Time')

# Chart 2: Bar - Subject-wise marks
axes[0, 1].bar(['Math', 'Phy', 'Chem'], [85, 78, 92])
axes[0, 1].set_title('Subject Marks')

# Chart 3: Pie - Time distribution
axes[1, 0].pie([30, 25, 20, 25], labels=['Study', 'Sleep', 'Play', 'Other'], autopct='%1.1f%%')
axes[1, 0].set_title('Time Distribution')

# Chart 4: Scatter - Study vs marks
axes[1, 1].scatter([1,2,3,4,5], [50,60,70,80,90])
axes[1, 1].set_title('Study vs Score')

plt.tight_layout()
plt.show()
```

## 7. Common Customizations

### Common Customization Functions

- `plt.figure(figsize=(w, h))` → Sets the graph size (width, height).
- `'ro-'` → `r` = red, `o` = circle markers, `-` = solid line.
- `label='Data'` → Gives the plotted data a name for the legend.
- `plt.title()` → Adds a title (`fontsize` changes size, `fontweight='bold'` makes it bold).
- `plt.xlabel()` → Labels the X-axis.
- `plt.ylabel()` → Labels the Y-axis.
- `plt.legend()` → Displays the legend.
- `plt.grid(True, alpha=0.3)` → Shows a light grid (`alpha` controls transparency).
- `plt.tight_layout()` → Adjusts spacing automatically.
- `plt.show()` → Displays the graph.

```python
plt.figure(figsize=(8, 4))   # Change figure size

x = range(5)
y = [3, 7, 2, 8, 5]

plt.plot(x, y, 'ro-', label='Data')  # r=red, o=circle, -=line
plt.title('Customized Chart', fontsize=16, fontweight='bold')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
