# Python Regular Expressions Masterclass
## Find, Match, and Replace Patterns in Text

Regular expressions (regex) let you search for patterns in text. They look scary at first but are just a small set of symbols you will memorize with practice.

## 1. Why Regex?

Imagine you have 10,000 customer records and need to:
- Find all invalid phone numbers
- Extract all email addresses
- Replace all dates from "DD/MM/YYYY" to "YYYY-MM-DD"

You CANNOT do this with `str.find()` or `str.replace()` alone. Regex is the right tool.

## 2. Your First Regex -- Finding a Simple Pattern

```python
# Import regular expressions module for pattern matching
import re

text = "My phone is 9876543210 and office is 1234567890"

# Find all 10-digit numbers
# {n} = exactly n repetitions
pattern = r"\d{10}"  # \d = digit, {10} = exactly 10 times
# Find ALL matches in text (returns list)
matches = re.findall(pattern, text)
print(matches)
```

## 3. The Most Important Regex Symbols

You can solve 80% of problems with just these:

| Symbol | Meaning | Example |
|--------|---------|---------|
| \\d | Any digit | \\d+ matches '123' |
| \\w | Any word char | \\w+ matches 'hello' |
| \\s | Any whitespace | \\s+ matches spaces |
| . | Any character | h.t matches 'hat' |
| + | One or more | \\d+ matches '5', '500' |
| * | Zero or more | \\d* matches '', '100' |
| ? | Zero or one | colou?r matches 'color' |
| {n} | Exactly n | \\d{10} = 10 digits |
| [abc] | Any of a,b,c | [aeiou] = any vowel |
| ^ | Start of string | ^Hello |
| $ | End of string | end$ |
| | | OR | cat|dog |
| () | Group | (\d{3})-(\d{4}) |

## 4. re.findall() -- Find All Matches

Returns a list of all matches. This is the function you will use most.

```python
text = "Emails: rahul@gmail.com, priya@yahoo.com, amit@outlook.com"

# Find all email addresses
pattern = r"[\w.]+@[\w.]+"
# Find ALL matches in text (returns list)
emails = re.findall(pattern, text)
print(emails)
```

## 5. re.search() -- Find First Match

```python
text = "Price: Rs.599 only. Offer: Rs.399."

# Find FIRST match anywhere in text (returns Match object)
match = re.search(r"Rs\.(\d+)", text)
if match:
# Get the full matched text
    print("First price found:", match.group())
# Get text from first capture group (parentheses)
    print("Just the number:", match.group(1))
```

## 6. re.sub() -- Find and Replace

```python
text = "My SSN is 123-45-6789 and my friend's is 987-65-4321"

# Replace SSN with ***
# Find and REPLACE all matches in text
masked = re.sub(r"\d{3}-\d{2}-\d{4}", "***-**-****", text)
print(masked)
```

```python
# Swap date format: DD/MM/YYYY -> YYYY-MM-DD
dates = "15/08/1947 and 26/01/1950"
# Find and REPLACE all matches in text
swapped = re.sub(r"(\d{2})/(\d{2})/(\d{4})", r"\3-\2-\1", dates)
print(swapped)
```

## 7. re.match() vs re.search()

```python
# re.match() only checks the BEGINNING of the string
# Match pattern from START of string only
print(re.match(r"\d+", "123abc"))   # Matches -- starts with digits
# Match pattern from START of string only
print(re.match(r"\d+", "abc123"))   # None -- doesn't START with digits

# re.search() checks ANYWHERE in the string
# Find FIRST match anywhere in text (returns Match object)
print(re.search(r"\d+", "abc123"))  # Matches -- digits found inside
```

## 8. Groups -- Extract Parts of a Match

```python
text = "Contact: Rahul, Phone: 9876543210, Age: 22"

# Find FIRST match anywhere in text (returns Match object)
match = re.search(r"Phone: (\d{10}), Age: (\d+)", text)
if match:
# Get the full matched text
    print("Full match:", match.group())
# Get text from first capture group (parentheses)
    print("Phone:", match.group(1))
    print("Age:", match.group(2))
```

## 9. Real-World Examples

```python
# Validate Indian phone number (10 digits, starts with 6-9)
def is_valid_phone(number):
    pattern = r"^[6-9]\d{9}$"
# Match pattern from START of string only
    return bool(re.match(pattern, number))

print(is_valid_phone("9876543210"))  # True
print(is_valid_phone("1234567890"))  # False - starts with 1
print(is_valid_phone("98765"))       # False - too short
```

```python
# Extract all hashtags from a tweet
tweet = "Loving #Python and #MachineLearning! #100DaysOfCode"
# Find ALL matches in text (returns list)
hashtags = re.findall(r"#(\w+)", tweet)
print(hashtags)
```

```python
# Check if string is a valid email
def is_valid_email(email):
# [chars] = character class (any one of these characters)
    pattern = r"^[\w.+-]+@[\w-]+\.[a-z]{2,}$"
# Match pattern from START of string only
    return bool(re.match(pattern, email))

print(is_valid_email("rahul@gmail.com"))  # True
print(is_valid_email("not-an-email"))     # False
```

## 10. When NOT to Use Regex

Regex is powerful but overused. Don't use it for:
- Simple substring checks -> use `'sub' in string`
- Exact string matching -> use `==`
- Simple prefix/suffix -> use `str.startswith()` / `str.endswith()`
- Parsing HTML/XML -> use BeautifulSoup

## 11. Practice Exercises

```python
# 1. Extract all numbers from a string
text = "Order #4521: 3 items, total Rs.1500"
# Find ALL matches in text (returns list)
numbers = re.findall(r"\d+", text)
print(numbers)

# 2. Check if password has at least 8 chars, 1 digit, 1 uppercase
def is_strong_password(pw):
    if len(pw) < 8: return False
# Find FIRST match anywhere in text (returns Match object)
    if not re.search(r"\d", pw): return False
# Find FIRST match anywhere in text (returns Match object)
    if not re.search(r"[A-Z]", pw): return False
    return True

print(is_strong_password("Hello123"))   # True
print(is_strong_password("hello"))      # False
```
