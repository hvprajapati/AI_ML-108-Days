# PART 1: datetime Module

## 1. Getting the Current Date and Time

```python
# Import datetime module for working with dates and times
from datetime import datetime, date, time, timedelta

# Get current date and time
now = datetime.now()
print("Full datetime:", now)
# Create a specific date (year, month, day)
print("Date only:", now.date())
print("Time only:", now.time())
print("Year:", now.year, "Month:", now.month, "Day:", now.day)
# Get day of week (0=Monday, 6=Sunday)
print("Weekday (0=Mon):", now.weekday())
```

## 2. Creating Specific Dates

```python
# Create a specific date
# Create a specific date (year, month, day)
independence_day = date(1947, 8, 15)
print("Independence Day:", independence_day)

# Create a specific datetime
# Create a specific datetime (year, month, day, hour, min, sec)
exam_time = datetime(2024, 3, 15, 10, 30, 0)
print("Exam starts at:", exam_time)
```

## 3. Formatting Dates (strftime)

Convert datetime -> string in ANY format you want.

```python
# Get current date and time
now = datetime.now()

# Common formats you will use daily
# Format datetime object to string (str = string, f = format, time)
print(now.strftime("%d/%m/%Y"))          # 28/06/2024
# Format datetime object to string (str = string, f = format, time)
print(now.strftime("%d-%b-%Y"))          # 28-Jun-2024
# Format datetime object to string (str = string, f = format, time)
print(now.strftime("%A, %d %B %Y"))      # Friday, 28 June 2024
# Format datetime object to string (str = string, f = format, time)
print(now.strftime("%I:%M %p"))           # 03:45 PM
# Format datetime object to string (str = string, f = format, time)
print(now.strftime("%Y-%m-%d %H:%M:%S"))  # ISO format
```

### strftime codes you will use most:

| Code | Meaning | Example |
|------|---------|--------|
| %d | Day (01-31) | 28 |
| %m | Month (01-12) | 06 |
| %Y | Year (4 digits) | 2024 |
| %b | Month abbreviation | Jun |
| %B | Month full name | June |
| %A | Weekday full name | Friday |
| %H | Hour (24h) | 15 |
| %M | Minute | 45 |
| %S | Second | 30 |
| %p | AM/PM | PM |

## 4. Parsing Dates from Strings (strptime)

Convert string -> datetime. Essential when reading dates from files or user input.

```python
date_str = "15-08-1947"
# Parse string to datetime (str = string, p = parse, time)
parsed = datetime.strptime(date_str, "%d-%m-%Y")
print(parsed)
# Format datetime object to string (str = string, f = format, time)
print(parsed.strftime("%A, %d %B %Y"))  # Now we can reformat it!

# Another example
log_entry = "2024-06-28 14:30:00"
# Parse string to datetime (str = string, p = parse, time)
log_time = datetime.strptime(log_entry, "%Y-%m-%d %H:%M:%S")
print("Hour of log:", log_time.hour)
```

## 5. Date Arithmetic (timedelta)

Add or subtract days, hours, weeks to/from dates.

```python
# Get today's date
today = date.today()
print("Today:", today)

# Common operations
# Create a time difference (days, hours, seconds)
print("Tomorrow:", today + timedelta(days=1))
# Create a time difference (days, hours, seconds)
print("Yesterday:", today - timedelta(days=1))
# Create a time difference (days, hours, seconds)
print("Next week:", today + timedelta(weeks=1))
# Create a time difference (days, hours, seconds)
print("30 days from now:", today + timedelta(days=30))

# Calculate age
# Create a specific date (year, month, day)
birth_date = date(2002, 5, 15)
age_days = (today - birth_date).days
print(f"Days alive: {age_days}")
print(f"Approx age: {age_days // 365} years")
```

## 6. Real-World Examples

```python
# Calculate days until a deadline
# Create a specific date (year, month, day)
deadline = date(2024, 12, 31)
# Get today's date
remaining = (deadline - date.today()).days
print(f"Days until deadline: {remaining}")

# Check if a date is a weekday or weekend
# Create a specific date (year, month, day)
d = date(2024, 6, 29)  # Saturday
# Get day of week (0=Monday, 6=Sunday)
if d.weekday() >= 5:
    print(f"{d} is a weekend!")
else:
    print(f"{d} is a weekday.")
```

# PART 2: random Module

## 7. Generating Random Numbers

```python
# Import random module for generating random numbers
import random

# Generate random integer between a and b (both included)
print("Random integer 1-10:", random.randint(1, 10))
# Generate random float between 0 and 1
print("Random float 0-1:", random.random())
# Generate random float between two values
print("Random float 1-100:", random.uniform(1, 100))
# Random integer from range(start, stop, step)
print("Random from range:", random.randrange(0, 100, 5))  # 0,5,10,...,95
```

## 8. Choosing Random Items

```python
colors = ["red", "blue", "green", "yellow", "purple"]

# Pick one
# Pick one random item from a sequence
print("Random choice:", random.choice(colors))

# Pick 3 (may repeat)
# Pick k random items (may repeat)
print("3 choices:", random.choices(colors, k=3))

# Pick 3 (unique, no repeats)
# Pick k unique random items (no repeats)
print("3 unique:", random.sample(colors, k=3))
```

## 9. Shuffling

```python
cards = ["Ace", "King", "Queen", "Jack", "10"]
print("Before:", cards)
# Randomly shuffle a list in-place
random.shuffle(cards)
print("After:", cards)
```

## 10. Setting a Seed (Reproducible Randomness)

Same seed -> same sequence. Essential for debugging and testing.

```python
# Set seed for reproducible random numbers
random.seed(42)
# Generate random integer between a and b (both included)
print(random.randint(1, 100))  # Always 82 with seed 42
# Generate random integer between a and b (both included)
print(random.randint(1, 100))  # Always 15 with seed 42

# Set seed for reproducible random numbers
random.seed(42)  # Reset
# Generate random integer between a and b (both included)
print(random.randint(1, 100))  # 82 again!
# Generate random integer between a and b (both included)
print(random.randint(1, 100))  # 15 again!
```
