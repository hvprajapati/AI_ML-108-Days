## 1. What is Type Conversion?
Type conversion means converting one data type into another data type.

```python
age = '22'
print(type(age))
age = int(age)
print(type(age))
```

## 2. Implicit Type Conversion
Python automatically converts smaller compatible types into larger types.

```python
a = 10
b = 5.5
result = a + b
print(result)
print(type(result))
```

## 3. Explicit Type Conversion
You manually convert types using functions.

```python
num = '100'
num = int(num)
print(num + 50)
```

## int() Conversion

```python
print(int('25'))
print(int(10.99))
print(int(True))
```

## float() Conversion

```python
print(float('25.5'))
print(float(10))
```

## str() Conversion

```python
salary = 50000
print(str(salary))
```

## bool() Conversion

```python
print(bool(1))
print(bool(0))
print(bool('Python'))
print(bool(''))
```

## list(), tuple(), set()

```python
text = 'python'
print(list(text))
print(tuple(text))
print(set(text))
```
