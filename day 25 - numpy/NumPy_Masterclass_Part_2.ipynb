## 2. Why Vectorization?

```python
# Import NumPy with standard alias np
import numpy as np
import time

# Create array with evenly spaced values (start, stop, step)
numbers = np.arange(1_000_000)

start = time.time()
result = numbers * 2
print("Vectorized Time:", time.time() - start)   

#if same thing would have done by the for loop it would have taken lot much time 
# so directly doing task without for loop is vectorization
```

Vectorization performs operations on an entire array without explicit Python loops.

## 3. Universal Functions (ufuncs)

```python
# Import NumPy with standard alias np
import numpy as np

# Create a NumPy array from a Python list
arr = np.array([1, 4, 9, 16])

print(np.sqrt(arr))
print(np.square(arr))
print(np.sin(arr))
print(np.cos(arr))
```

## 4. Aggregation Functions

```python
# Import NumPy with standard alias np
import numpy as np

# Create a NumPy array from a Python list
arr = np.array([10,20,30,40,50])

print(np.sum(arr))
print(np.mean(arr))
print(np.median(arr))
print(np.std(arr))
print(np.var(arr))
print(np.min(arr))
print(np.max(arr))
print(np.argmin(arr))
print(np.argmax(arr))
```

### `np.min(arr)`
Returns the **smallest value** in the array.

### `np.max(arr)`
Returns the **largest value** in the array.

### `np.argmin(arr)`
Returns the **index (position)** of the smallest value in the array.

### `np.argmax(arr)`
Returns the **index (position)** of the largest value in the array.

## 5. Understanding the `axis` Parameter

### `axis` in NumPy
- `axis=0` → Perform the operation **column-wise** (down the rows).
- `axis=1` → Perform the operation **row-wise** (across the columns).

```python
# Create a NumPy array from a Python list
matrix = np.array([[1,2,3],
                   [4,5,6]])

print("Sum of all:", matrix.sum())
print("Column sums:", matrix.sum(axis=0))
print("Row sums:", matrix.sum(axis=1))
```

## 6. Boolean Masking

```python
# Create a NumPy array from a Python list
arr = np.array([12,5,30,7,42,18])

mask = arr > 15

print(mask)
# Index array by position (0-based)
print(arr[mask])
```

## 7. Fancy Indexing

```python
# Create a NumPy array from a Python list
arr = np.array([10,20,30,40,50,60])

# Index array by position (0-based)
print(arr[[0,2,5]])


# Index:  0   1   2   3   4   5
# Value: 10  20  30  40  50  60
#         ↑       ↑           ↑
```

## 8. Sorting & Searching

```python
# Create a NumPy array from a Python list
arr = np.array([40,10,30,20])

print(np.sort(arr))
print(np.argsort(arr))
# Returns the indices that would sort the array.
```

## 9. Array Manipulation

```python
# Create a NumPy array from a Python list
a = np.array([1,2,3])
# Create a NumPy array from a Python list
b = np.array([4,5,6])

print(np.concatenate((a,b)))
print(np.stack((a,b)))
print(np.hstack((a,b)))
print(np.vstack((a,b)))
```

## 11. Useful NumPy Utilities

### `np.unique(arr)`
Returns the **unique values** in the array (duplicates are removed).

### `np.clip(arr, min, max)`
Limits all values to the given range. Values below `min` become `min`, and values above `max` become `max`.

### `np.cumsum(arr)`
Returns the **cumulative (running) sum** of the array.

### `np.cumprod(arr)`
Returns the **cumulative (running) product** of the array.

```python
# Create a NumPy array from a Python list
arr = np.array([5,2,8,1])
# Returns only the unique values (duplicates removed).

print(np.unique(arr))

# Keeps every value between 2 and 6.
# Values less than 2 become 2.
# Values greater than 6 become 6.

print(np.clip(arr,2,6))

print(np.cumsum(arr))

print(np.cumprod([1,2,3,4]))
```
