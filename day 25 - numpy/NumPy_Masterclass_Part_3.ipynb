## 1. Linear Algebra

```python
# Import NumPy with standard alias np
import numpy as np

# Create a NumPy array from a Python list
A = np.array([[1,2],[3,4]])
# Create a NumPy array from a Python list
B = np.array([[5,6],[7,8]])

print("Dot Product:\n", np.dot(A,B))
print("Matrix Multiplication:\n", A @ B)
# The @ operator is just a shortcut for matrix multiplication.
```

```python
# Import NumPy with standard alias np
import numpy as np

# Create a NumPy array from a Python list
A = np.array([[4,7],[2,6]])

print("Determinant:", np.linalg.det(A))
print("Inverse:\n", np.linalg.inv(A))
# Transpose array (swap rows and columns)
print("Transpose:\n", A.T)
```

```python
# Import NumPy with standard alias np
import numpy as np

# Create a NumPy array from a Python list
A = np.array([[4,2],[1,3]])

values, vectors = np.linalg.eig(A)

print("Eigenvalues:", values)
print("Eigenvectors:\n", vectors)
```

## 2. Random Module

```python
# Import NumPy with standard alias np
import numpy as np

np.random.seed(42)

# Generate random floats between 0 and 1
print(np.random.rand(3))
# Generate random integers in a range
print(np.random.randint(1,100,5))
print(np.random.choice([10,20,30], size=5))
print(np.random.permutation([1,2,3,4]))
```

```python
# Import NumPy with standard alias np
import numpy as np

# Create array with evenly spaced values (start, stop, step)
data=np.arange(10)
np.random.shuffle(data)
print(data)
```

## 3. Statistical Operations

```python
# Import NumPy with standard alias np
import numpy as np

# Create a NumPy array from a Python list
arr=np.array([12,18,20,25,30])

print("Mean:", np.mean(arr))
print("Median:", np.median(arr))
print("Std:", np.std(arr))
print("Variance:", np.var(arr))
print("Percentile:", np.percentile(arr,75))
```

## 4. Correlation

### `np.corrcoef(x, y)`
Returns the **correlation coefficient matrix**, which measures how strongly two arrays are related. A value of `1` means perfect positive correlation, `0` means no correlation, and `-1` means perfect negative correlation.

```python
# Import NumPy with standard alias np
import numpy as np

# Create a NumPy array from a Python list
x=np.array([1,2,3,4,5])
# Create a NumPy array from a Python list
y=np.array([2,4,6,8,10])

print(np.corrcoef(x,y))
```

## 5. NumPy File I/O

```python
# Import NumPy with standard alias np
import numpy as np

# Create array with evenly spaced values (start, stop, step)
arr=np.arange(10)

np.save("array.npy",arr)

loaded=np.load("array.npy")

print(loaded)
```

```python
# Import NumPy with standard alias np
import numpy as np

# Create array with evenly spaced values (start, stop, step)
arr=np.arange(12).reshape(3,4)

np.savetxt("matrix.txt",arr,fmt="%d")

loaded=np.loadtxt("matrix.txt")

print(loaded)
```

## 6. Structured Arrays

### Structured `dtype`
A structured `dtype` lets each row contain **multiple fields with different data types**, similar to a table or database record. Here, `"name"` is a Unicode string (`U10`) and `"marks"` is a 4-byte integer (`i4`).

```python
# Import NumPy with standard alias np
import numpy as np

# Create a NumPy array from a Python list
students=np.array([
    ("Alice",90),
    ("Bob",82),
    ("Charlie",95)
# Specify data type when creating array
],dtype=[("name","U10"),("marks","i4")])

print(students)
print(students["name"])
print(students["marks"])
```

## 10. Real-World Project - Exam Analytics

```python
# Import NumPy with standard alias np
import numpy as np

# Create a NumPy array from a Python list
marks=np.array([
    [80,90,70],
    [75,88,92],
    [95,91,89]
])

print("Student Average:",marks.mean(axis=1))
print("Subject Average:",marks.mean(axis=0))
# Find the maximum value in the array
print("Highest:",marks.max())
# Find the minimum value in the array
print("Lowest:",marks.min())
```
