## 4. Creating Arrays

```python
# Import NumPy with standard alias np
import numpy as np

print(np.array([1,2,3]))
# Create array filled with zeros
print(np.zeros(5))
# Create array filled with ones
print(np.ones(5))
# Create array filled with a specific value
print(np.full(5,100))
# Create uninitialized array (random values from memory)
print(np.empty(3)) # It creates an array by allocating memory without initializing its values.
```

```python
print(np.arange(1,11))
# Creates numbers from 1 to 10 (11 is excluded).
print(np.arange(0,20,2))
# Creates numbers from 0 to 18 with a step of 2.
print(np.linspace(0,1,5))
# Creates 5 evenly spaced numbers between 0 and 1 (both included).
```

```python
# Create identity matrix (1s on diagonal, 0s elsewhere)
print(np.eye(4))
# Create square identity matrix
print(np.identity(3))
```

```python
print(np.identity(3,5))
```

```python
print(np.eye(3, 5))
```

## 5. Random Arrays

```python
# Import NumPy with standard alias np
import numpy as np

# Generate random floats between 0 and 1
print(np.random.rand(3))
# Generate random integers in a range
print(np.random.randint(1,100,5))
# Generate random numbers from standard normal distribution
print(np.random.randn(5))
```

## 6. Array Attributes

```python
# Create a NumPy array from a Python list
arr=np.array([[1,2,3],[4,5,6]])

print(arr)
# Print array shape (rows, columns)
print("Shape:",arr.shape)
# Print number of dimensions
print("Dimensions:",arr.ndim)
# Print total number of elements
print("Size:",arr.size)
# Print data type of array elements
print("Data Type:",arr.dtype)
```

## 7. Data Types

```python
# Create a NumPy array from a Python list
arr=np.array([1,2,3],dtype=float)
print(arr)
# Print data type of array elements
print(arr.dtype)

# Convert array to a different data type
print(arr.astype(int))
```

## 8. Indexing

```python
# Create a NumPy array from a Python list
arr=np.array([10,20,30,40,50])

# Index array by position (0-based)
print(arr[0])
# Index array by position (0-based)
print(arr[-1])
```

```python
# Create a NumPy array from a Python list
matrix=np.array([[1,2,3],[4,5,6]])

print(matrix[0,1])
print(matrix[1,2])
```

## 9. Slicing

```python
# Create array with evenly spaced values (start, stop, step)
arr=np.arange(10)

# Slice array [start:stop:step]
print(arr[2:7])
# Slice array [start:stop:step]
print(arr[:5])
# Slice array [start:stop:step]
print(arr[::2])
# Slice array [start:stop:step]
print(arr[::-1])
```

```python
# Create array with evenly spaced values (start, stop, step)
matrix=np.arange(1,10).reshape(3,3)

print(matrix)
print("*"*50)
print(matrix[:,:])
print("*"*50)

print(matrix[:,1])
print("*"*50)

print(matrix[1,:])
print("*"*50)

print(matrix[0:2,0:2])
```

## 10. Reshaping Arrays

```python
# Create array with evenly spaced values (start, stop, step)
arr=np.arange(12)

# Reshape array to new dimensions (total elements must match)
print(arr.reshape(3,4))
# Reshape array to new dimensions (total elements must match)
print(arr.reshape(2,6))
```

```python
# Create array with evenly spaced values (start, stop, step)
arr=np.arange(6).reshape(2,3)

# Flatten multi-dimensional array to 1D (returns copy)
print(arr.flatten())
# Flatten to 1D (returns view when possible, faster)
print(arr.ravel())
# Transpose array (swap rows and columns)
print(arr.T)
```

## 11. Copy vs View

```python
# Create a NumPy array from a Python list
arr=np.array([1,2,3])

# Create a view (shares data with original)
view=arr.view()
# Create a deep copy (independent data)
copy=arr.copy()

arr[0]=999

print(arr)
print(view)
print(copy)
```

# Copy vs View

- **Copy**: Makes a completely new array with its own data.
- Changes to the copy do not affect the original.
- Changes to the original do not affect the copy.

- **View**: Makes a new array object that uses the same data as the original.
- Changes to the view affect the original.
- Changes to the original affect the view.

## 12. Basic Mathematical Operations

```python
# Create a NumPy array from a Python list
a=np.array([1,2,3])
# Create a NumPy array from a Python list
b=np.array([4,5,6])

# Element-wise operation between two arrays
print(a+b)
# Element-wise operation between two arrays
print(a-b)
# Element-wise operation between two arrays
print(a*b)
# Element-wise operation between two arrays
print(a/b)
# Element-wise power (square each element)
print(a**2)
```

## 13. Broadcasting Basics

```python
# Create a NumPy array from a Python list
arr=np.array([1,2,3])

# Broadcasting: operation applied to every element
print(arr+10)
# Broadcasting: operation applied to every element
print(arr*5)
```
