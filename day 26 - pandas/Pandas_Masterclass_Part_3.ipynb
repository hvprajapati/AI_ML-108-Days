## 1. Date & Time with Pandas

```
# Import Pandas with standard alias pd
import pandas as pd

dates = pd.DataFrame({
    "OrderDate":["2026-01-10","2026-02-15","2026-03-20"]
})

# Convert column to datetime format
dates["OrderDate"] = pd.to_datetime(dates["OrderDate"])

print(dates)
print(dates["OrderDate"].dt.year)
print(dates["OrderDate"].dt.month_name())
print(dates["OrderDate"].dt.dayofweek)
```

## 2. Creating a Time Series

```
# Import Pandas with standard alias pd
import pandas as pd
# Import NumPy with standard alias np
# Import NumPy with standard alias np
import numpy as np

# Generate random integers in a range
ts = pd.Series(np.random.randint(100,200,10),
               index=pd.date_range("2026-01-01", periods=10))

print(ts)
# Calculate the mean (average) of array elements
print(ts.resample("5D").mean())
```

## 3. MultiIndex

```
# Import Pandas with standard alias pd
import pandas as pd

# Create/modify DataFrame
df = pd.DataFrame({
    "Department":["IT","IT","HR","HR"],
    "Year":[2025,2026,2025,2026],
    "Salary":[70000,75000,60000,65000]
})

# Set a column as the DataFrame index
multi = df.set_index(["Department","Year"])
print(multi)
# Select rows/columns by label name
print(multi.loc[("IT",2026)])
```

## 4. Window Functions

```
# Import Pandas with standard alias pd
import pandas as pd

sales = pd.Series([100,120,150,180,170,200])

# Calculate the mean (average) of array elements
print(sales.rolling(3).mean())
# Calculate the mean (average) of array elements
print(sales.expanding().mean())
```

## 8. SQL Example

```
import sqlite3
# Import Pandas with standard alias pd
import pandas as pd

conn = sqlite3.connect(":memory:")

# Create/modify DataFrame
df = pd.DataFrame({
    "Name":["Alice","Bob"],
    "Salary":[70000,65000]
})

df.to_sql("employees", conn, index=False)

result = pd.read_sql("SELECT * FROM employees", conn)
print(result)
```

## 10. Real-World Project – Sales Dashboard

```
# Import Pandas with standard alias pd
import pandas as pd
# Import NumPy with standard alias np
# Import NumPy with standard alias np
import numpy as np

np.random.seed(42)

sales = pd.DataFrame({
    "Region":np.random.choice(["North","South","East","West"],100),
    "Product":np.random.choice(["Laptop","Phone","Tablet"],100),
# Generate random integers in a range
    "Sales":np.random.randint(1000,10000,100)
})

# Display first 5 rows of the DataFrame
print(sales.head())

print("\nSales by Region")
# Group data by a column and apply aggregation
print(sales.groupby("Region")["Sales"].sum())

print("\nSales by Product")
# Calculate the mean (average) of array elements
# Group data by a column and apply aggregation
print(sales.groupby("Product")["Sales"].mean())

# Create pivot table (like Excel pivot)
pivot = pd.pivot_table(
    sales,
    values="Sales",
    index="Region",
    columns="Product",
    aggfunc="sum"
)

print("\nPivot Table")
print(pivot)
```

## 11. Performance Tips


- Prefer vectorized operations.
- Avoid looping through rows.
- Use appropriate data types.
- Read only required columns (`usecols`).
- Use `categorical` data for repeated strings.


## 12. Common Mistakes


- Forgetting `parse_dates`.
- Using loops instead of Pandas operations.
- Confusing `loc` and `iloc`.
- Ignoring missing values before analysis.
- Loading unnecessary columns into memory.
