# 1. Preparing a Realistic Dataset

We'll create a dataset with missing values and duplicates to practice real-world data cleaning.


```
# Import Pandas with standard alias pd
import pandas as pd
# Import NumPy with standard alias np
# Import NumPy with standard alias np
import numpy as np

np.random.seed(42)

# Create/modify DataFrame
df = pd.DataFrame({
    "EmployeeID":[1001,1002,1003,1004,1005,1006,1006],
    "Department":["IT","HR","IT","Sales","Finance","IT","IT"],
    "Salary":[75000,62000,np.nan,58000,91000,83000,83000],
    "Experience":[5,2,7,3,np.nan,8,8],
    "Rating":[4.5,4.0,np.nan,3.8,4.9,4.7,4.7]
})

df.to_csv("employees_dirty.csv",index=False)
print(df)
```

## 2. Loading the Dataset

```
# Import Pandas with standard alias pd
import pandas as pd
# Load data from a CSV file into a DataFrame
df=pd.read_csv("employees_dirty.csv")
print(df)
```

## 3. Missing Values

```
print(df.isnull())
# Count missing values in each column
print(df.isnull().sum())
```

```
# Remove rows with missing values
print(df.dropna())
```

```
print(df.fillna(0))
# Calculate the mean (average) of array elements
print(df.fillna(df["Salary"].mean()))
```

## 4. Duplicate Records

```
# Check for duplicate rows
print(df.duplicated())
# Remove duplicate rows
print(df.drop_duplicates())
```

## 5. Replacing Values

```
# Replace values in DataFrame
print(df.replace("IT","Information Technology"))
```

## 6. String Operations

```
# Access string methods on a Series
dept=df["Department"].str.upper()
print(dept)

# Access string methods on a Series
print(df["Department"].str.contains("IT"))
```

## 7. GroupBy

```
# Calculate the mean (average) of array elements
print(df.groupby("Department")["Salary"].mean())
```

```
# Group data by a column and apply aggregation
print(df.groupby("Department").agg({
    "Salary":["mean","max","min"],
    "Experience":"mean"
}))
```

## 8. Sorting & Ranking

```
# Sort DataFrame by specified column
print(df.sort_values("Salary",ascending=False))
print(df["Salary"].rank())
```

## 9. Merge DataFrames

```
employees=pd.DataFrame({
    "EmployeeID":[1001,1002,1003],
    "Name":["Alice","Bob","Charlie"]
})

salary=pd.DataFrame({
    "EmployeeID":[1001,1002,1003],
    "Salary":[70000,60000,90000]
})

# Merge two DataFrames (like SQL JOIN)
merged=pd.merge(employees,salary,on="EmployeeID")
print(merged)
```

## 10. Concatenation

```
a=pd.DataFrame({"A":[1,2]})
b=pd.DataFrame({"A":[3,4]})

# Concatenate DataFrames along rows or columns
print(pd.concat([a,b],ignore_index=True))
```

## 11. Pivot Tables

```
sales=pd.DataFrame({
    "Region":["North","North","South","South"],
    "Product":["Laptop","Phone","Laptop","Phone"],
    "Sales":[500,300,700,400]
})

# Create pivot table (like Excel pivot)
print(pd.pivot_table(sales,
                     values="Sales",
                     index="Region",
                     columns="Product",
                     aggfunc="sum"))
```

## 12. Crosstab

```
print(pd.crosstab(sales["Region"],sales["Product"]))
```

## 13. Apply Functions

```
# Calculate the mean (average) of array elements
# Fill missing values with a specified value
df["Salary_Filled"]=df["Salary"].fillna(df["Salary"].mean())
# Apply a function to each element/row/column
df["Tax"]=df["Salary_Filled"].apply(lambda x:x*0.1)
print(df)
```

## 14. Mini Project - Department Analytics

```
# Fill missing values with a specified value
clean=df.fillna(df.mean(numeric_only=True))

# Calculate the mean (average) of array elements
# Group data by a column and apply aggregation
print(clean.groupby("Department")[["Salary","Experience"]].mean())
print("\nHighest Salary Department")
# Find the maximum value in the array
# Group data by a column and apply aggregation
print(clean.groupby("Department")["Salary"].max())
```
