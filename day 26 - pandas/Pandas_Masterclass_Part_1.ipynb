## 2. Create a Custom Dataset

```python
import pandas as pd

# Create a sample employee dataset
df = pd.DataFrame({
    "EmployeeID": [1001, 1002, 1003, 1004, 1005],
    "Name": ["Alice", "Bob", "Charlie", "David", "Emma"],
    "Department": ["IT", "HR", "Finance", "Sales", "Marketing"],
    "Age": [25, 30, 28, 35, 27],
    "Salary": [60000, 50000, 75000, 80000, 65000],
    "Experience": [2, 5, 4, 8, 3],
    "Rating": [4.5, 3.8, 4.7, 4.2, 4.9]
})

# Save the DataFrame as a CSV file
df.to_csv("employees.csv", index=False)

print("employees.csv created successfully!")
print(df)
```

From now on, every example in this notebook uses **employees.csv** that you just created.

You can also replace it later with any real dataset.


## 3. Reading CSV Files

```python
# Load data from a CSV file into a DataFrame
df = pd.read_csv("employees.csv")

# Display first 5 rows of the DataFrame
print(df.head())
```

## 4. Understanding DataFrames & Series

```python
print(type(df))
print(type(df["Salary"]))
```

## 5. Basic Inspection

```python
# Display first 5 rows of the DataFrame
print(df.head())
# Display last 5 rows of the DataFrame
print(df.tail())
print(df.sample(5))
# Print array shape (rows, columns)
# Show DataFrame dimensions (rows, columns)
print(df.shape)
# Show all column names
print(df.columns)
print(df.index)
# Print data type of array elements
# Show data type of each column
print(df.dtypes)
```

## 6. Dataset Information

```python
# Show column names, data types, and non-null counts
print(df.info())
print(df.describe(include='all'))
```

## 7. Selecting Data

```python
print(df["Name"])
print(df[["Name","Salary"]])
```

```python
# Select rows/columns by label name
print(df.loc[0])
# Select rows/columns by label name
print(df.loc[0:4,["Name","Salary"]])
```

```python
# Select rows/columns by integer position
print(df.iloc[0])
# Select rows/columns by integer position
print(df.iloc[0:5,1:4])
```

## 8. Adding New Columns

```python
df["Bonus"] = df["Salary"] * 0.10
df["TotalSalary"] = df["Salary"] + df["Bonus"]

# Display first 5 rows of the DataFrame
print(df.head())
```

## 9. Renaming Columns

```python
# Rename columns or index labels
df = df.rename(columns={"Rating":"PerformanceRating"})
# Show all column names
print(df.columns)
```

## 10. Dropping Columns

```python
# Drop specified rows from DataFrame
temp = df.drop(columns=["Bonus"])
# Display first 5 rows of the DataFrame
print(temp.head())
```

## 11. Sorting Data

```python
# Display first 5 rows of the DataFrame
print(df.sort_values("Salary").head())
# Display first 5 rows of the DataFrame
print(df.sort_values("Salary",ascending=False).head())
```

## 12. Filtering Data

```python
high_salary = df[df["Salary"] > 80000]
print(high_salary)

it_staff = df[df["Department"]=="IT"]
print(it_staff)
```

## 13. Value Counts

```python
# Count unique values in a column
print(df["Department"].value_counts())
```

## 14. Mini Project - HR Dashboard

```python
# Calculate the mean (average) of array elements
print("Average Salary:", df["Salary"].mean())
# Find the maximum value in the array
print("Highest Salary:", df["Salary"].max())
# Find the minimum value in the array
print("Lowest Salary:", df["Salary"].min())
print("\nDepartment Counts")
print(df["Department"].value_counts())
```
