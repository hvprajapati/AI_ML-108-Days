## Why Do We Need Object-Oriented Programming?

Imagine you're building a library management system. You have books — each has a title, author, price. Without OOP, you'd use separate lists:

```python
titles = ["The Alchemist", "1984", "Clean Code"]
authors = ["Paulo Coelho", "George Orwell", "Robert Martin"]
prices = [299, 249, 599]
```

This works... until you need to add a new book, delete one, or search. The data for ONE book is scattered across THREE lists. One mistake in ordering and everything breaks.

**OOP bundles related data and behavior into a single unit — a CLASS.**

## 1. What is a Class?

A **class** is a BLUEPRINT. It defines what properties (data) and methods (behavior) every object of that type will have.

An **object** is an INSTANCE created from the blueprint.

```
Class = Blueprint of a house (the plan/design)
Object = The actual house built from that blueprint
```

You can build MANY houses from ONE blueprint. Same with classes — one Book class, many book objects.

## 3. Understanding `self` — The Most Important Word in OOP

`self` refers to THE SPECIFIC OBJECT that is calling the method.

When you write `book1.title = "The Alchemist"`, Python internally does: `Book.title = "The Alchemist"` in the context of `book1`.

**Think of `self` as "my"** — `self.title` means "MY title", `self.price` means "MY price".

```python
book1.title  # "MY (book1's) title"
book2.title  # "MY (book2's) title"
```

## 4. Adding Behavior — Methods

````python
class Book:
    def __init__(self, title, author, price):
        self.title = title
        self.author = author
        self.price = price
        self.is_borrowed = False  # Every book starts as available

    def borrow(self):
        # Method that CHANGES the object's state
        if self.is_borrowed:
            print(f"'{self.title}' is already borrowed!")
        else:
            self.is_borrowed = True
            print(f"'{self.title}' has been borrowed.")

    def return_book(self):
        # Method that restores state
        if not self.is_borrowed:
            print(f"'{self.title}' is already available!")
        else:
            self.is_borrowed = False
            print(f"'{self.title}' has been returned.")

    def get_info(self):
        # Method that returns information about the object
        status = "Borrowed" if self.is_borrowed else "Available"
        return f"'{self.title}' by {self.author} — Rs.{self.price} — {status}"

# Create and use objects
b = Book("Atomic Habits", "James Clear", 399)
print(b.get_info())     # Available
b.borrow()              # Atomic Habits has been borrowed.
print(b.get_info())     # Borrowed
b.return_book()         # Atomic Habits has been returned.
````

## 5. Instance Variables vs Class Variables

**Instance variable:** Unique to EACH object. Defined with `self.` inside `__init__`.
**Class variable:** SHARED across ALL objects. Defined outside `__init__`, directly in the class.

````python
class Student:
    # Class variable — shared by ALL students
    school = "International School"
    total_students = 0

    def __init__(self, name, grade):
        # Instance variables — unique to each student
        self.name = name
        self.grade = grade
        Student.total_students += 1  # Access class variable with ClassName.variable

    def get_info(self):
        return f"{self.name} studies in grade {self.grade} at {Student.school}"

# Create students
s1 = Student("Rahul", "10th")
s2 = Student("Priya", "12th")
s3 = Student("Amit", "9th")

print(s1.get_info())
print(s2.get_info())
print(f"\nAll go to: {Student.school}")         # Same school for everyone
print(f"Total students enrolled: {Student.total_students}")  # 3
````

## 6. The `__str__` Method — Make Print Beautiful

When you `print(book)`, Python calls `book.__str__()`. By default, you get `<__main__.Book object at 0x...>`. Override `__str__` to make it meaningful.

````python
class Book:
    def __init__(self, title, author, price):
        self.title = title
        self.author = author
        self.price = price

    def __str__(self):
        # Called by print() and str()
        return f"'{self.title}' by {self.author} (Rs.{self.price})"

    def __repr__(self):
        # Called by repr() — should be unambiguous, ideally recreatable
        return f"Book(title='{self.title}', author='{self.author}', price={self.price})"

b1 = Book("Sapiens", "Yuval Noah Harari", 599)
print(b1)           # Uses __str__: 'Sapiens' by Yuval Noah Harari (Rs.599)
print(repr(b1))     # Uses __repr__: Book(title='Sapiens', author='Yuval Noah Harari', price=599)
````

## 7. Complete Example — A Working Book Class

````python
class Book:
    # Class variable
    total_books = 0

    def __init__(self, title, author, price):
        self.title = title
        self.author = author
        self.price = price
        self.is_borrowed = False
        Book.total_books += 1  # Increment count every time a book is created

    def borrow(self):
        if not self.is_borrowed:
            self.is_borrowed = True
            return f"Borrowed: {self.title}"
        return f"Sorry, '{self.title}' is already borrowed"

    def return_book(self):
        if self.is_borrowed:
            self.is_borrowed = False
            return f"Returned: {self.title}"
        return f"'{self.title}' is already in the library"

    def __str__(self):
        status = "Available" if not self.is_borrowed else "Borrowed"
        return f"[{status}] '{self.title}' by {self.author} — Rs.{self.price}"

# Build a mini library
library = [
    Book("The Alchemist", "Paulo Coelho", 299),
    Book("1984", "George Orwell", 249),
    Book("Clean Code", "Robert Martin", 599),
    Book("Sapiens", "Yuval Noah Harari", 499),
]

# Use the library
print(library[0].borrow())      # Borrowed: The Alchemist
print(library[1].borrow())      # Borrowed: 1984
print(library[0].borrow())      # Sorry, already borrowed

print(f"\nLibrary Status:")
for book in library:
    print(f"  {book}")

print(f"\nTotal books created: {Book.total_books}")
````

## Key Takeaways

- **Class** = blueprint. **Object** = instance created from blueprint.
- **`__init__`** = constructor, runs when object is created.
- **`self`** = refers to the current object ("my" data).
- **Instance variables** = unique per object (`self.name`).
- **Class variables** = shared across all objects (`ClassName.var`).
- **`__str__`** = controls what `print()` displays.
- Methods define what objects can DO.
