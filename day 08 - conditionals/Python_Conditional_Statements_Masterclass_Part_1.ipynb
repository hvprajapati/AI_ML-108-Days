# 1. Why Do We Need Decision Making?

Imagine an ATM.

- If the PIN is correct → Allow access.
- Otherwise → Reject access.

Or a weather app:

- If it is raining → Carry an umbrella.
- Otherwise → Enjoy your day.

A program often needs to **choose between different actions**. This is called **decision making**.


# 2. What is a Conditional Statement?

A conditional statement checks whether a condition is **True** or **False**.

```
Start
  │
  ▼
Condition
  │
 ┌───────┐
 │       │
True   False
 │       │
 ▼       ▼
Execute  Execute
Block A  Block B
```


# 3. Boolean Values

````python
print(True)
print(False)

print(type(True))
print(type(False))
````

# 4. Comparison Operators

````python
a = 10
b = 20

print(a == b)
print(a != b)
print(a > b)
print(a < b)
print(a >= b)
print(a <= b)
````

| Operator | Meaning |
|----------|---------|
| `==` | Equal |
| `!=` | Not Equal |
| `>` | Greater Than |
| `<` | Less Than |
| `>=` | Greater Than or Equal |
| `<=` | Less Than or Equal |

# 5. Your First `if` Statement

````python
age = 20

if age >= 18:
    print("You are eligible to vote.")
````

If the condition is **True**, the indented block executes. Otherwise, it is skipped.

# 6. `if-else` Statement

````python
age = 15

if age >= 18:
    print("Adult")
else:
    print("Minor")
````

# 7. `if-elif-else` Statement

````python
marks = 82

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 60:
    print("Grade C")
else:
    print("Grade D")
````

# 8. Nested `if`

````python
age = 22
citizen = False

if age >= 18:
    if citizen:
        print("Eligible to Vote")
    else:
        print("Not a citizen")
else:
    print("Underage")
````

# 9. Logical Operators

````python
age = 25
salary = 60000

if age >= 18 and salary >= 50000:
    print("Eligible for Premium Card")

if age < 18 or salary >= 50000:
    print("At least one condition is True")

is_logged_in = True

if not is_logged_in:
    print("Please login")
else:
    print("Welcome back!")
````

### Logical Operators

- `and` → Both conditions must be True.
- `or` → At least one condition must be True.
- `not` → Reverses the result.


# 10. Real-world Examples

````python
number = 17

if number % 2 == 0:
    print("Even")
else:
    print("Odd")
````

````python
year = 2024

if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print("Leap Year")
else:
    print("Not a Leap Year")
````

````python
a = 25
b = 40
c = 18

largest = a

if b > largest:
    largest = b

if c > largest:
    largest = c

print("Largest:", largest)
````

# 11. Common Mistakes


- Forgetting the colon (`:`)
- Wrong indentation
- Using `=` instead of `==`
- Forgetting that conditions must evaluate to `True` or `False`
- Incorrect nesting
