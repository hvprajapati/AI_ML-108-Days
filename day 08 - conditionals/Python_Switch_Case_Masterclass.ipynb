# 📘 Python Switch Case Masterclass

## From Traditional Switch Statements to Python's `match-case`

> **Prerequisite:** Python 3.10+

This notebook explains everything about **switch-case in Python**, including why Python originally didn't have one, the old alternatives, and the modern `match-case` statement with practical examples.


# 1. What is a Switch Statement?

Many programming languages like C, Java, and JavaScript have a **switch statement**.

Instead of writing many `if-elif-else` conditions, a switch statement selects one matching case.

### Traditional Example

```text
Input = 2

Switch
 │
 ├── Case 1
 ├── Case 2  ✅
 ├── Case 3
 └── Default
```


# 2. Did Python Always Have Switch Case?

❌ No.

Before Python 3.10, developers used:
- `if-elif-else`
- Dictionary Mapping
- Function Mapping

Python 3.10 introduced the **`match-case`** statement.


## 3. Using `if-elif-else`

````python
day = 3

if day == 1:
    print("Monday")
elif day == 2:
    print("Tuesday")
elif day == 3:
    print("Wednesday")
else:
    print("Invalid Day")
````

````python
days = {
    1: "Monday",
    2: "Tuesday",
    3: "Wednesday"
}

print(days.get(2, "Invalid Day"))
````

## 5. Your First `match-case`

````python
day = 5

match day:
    case 1:
        print("Monday")
    case 2:
        print("Tuesday")
    case 3:
        print("Wednesday")
    case _:
        print("Invalid Day")
````

### Understanding `case _`

`_` is the **default case**.

It is similar to:

- `default` in Java
- `default` in C


## 6. Matching Strings

````python
command = "start"

match command:
    case "start":
        print("Starting...")
    case "stop":
        print("Stopping...")
    case "restart":
        print("Restarting...")
    case _:
        print("Unknown Command")
````

## 7. Matching Multiple Values

````python
status = 403

match status:
    case 200:
        print("Success")
    case 400 | 401 | 403:
        print("Client Error")
    case 404:
        print("Not Found")
    case 500 | 501 | 503:
        print("Server Error")
    case _:
        print("Unknown")
````

## 8. Variable Capture

````python
value = ("Hardik", 24)

match value:
    case (name, age):
        print(name)
        print(age)
````

## 12. Mini Project - Simple Calculator

````python
choice = "+"

a = 10
b = 5

match choice:
    case "+":
        print(a + b)
    case "-":
        print(a - b)
    case "*":
        print(a * b)
    case "/":
        print(a / b)
    case _:
        print("Invalid Operation")
````

## 13. Common Mistakes

- Using `match-case` in Python versions older than **3.10**
- Forgetting the `:` after `match` or `case`
- Forgetting the default `case _`
- Thinking `match-case` replaces every `if` statement (it doesn't)


## 14. When Should You Use `match-case`?

✅ Fixed choices (commands, menu options, HTTP status codes)

❌ Complex boolean expressions (prefer `if-elif-else`)
