## 2. Short-hand `if`

```python
age = 20

if age >= 18:
    print("Adult")

# One-line version
if age >= 18: print("Adult")
```

## 3. Ternary (Conditional) Expression

```python
age = 16

status = "Adult" if age >= 18 else "Minor"
print(status)
```

## 4. Truthy vs Falsy Values

```python
values = [0, 1, "", "Python", [], [1], {}, {"a":1}, None, False, True]

for value in values:
    print(repr(value), "=>", bool(value))
```


### Common Falsy Values
- `False`
- `None`
- `0`
- `0.0`
- `""`
- `[]`
- `{}`
- `()`
- `set()`

Everything else is generally Truthy.


## 5. Identity Operators (`is` and `is not`)

```python
a = 5,4
b = 5,4
print(id(a))
print(id(b))
print(a is b)
print(a is not b)

x = "1,2"
y = "1,2"

print(x == y)   # Same values
print(x is y)   # Same object?

print(id(x))
print(id(y))
```

## 6. Membership Operators (`in` and `not in`)

```python
fruits = ["Apple", "Banana", "Mango"]

print("Apple" in fruits)
print("Orange" not in fruits)
```

## 7. Operator Precedence

```python
print(True or False and False)
print((True or False) and False)
```

```python
# | A     | B     | A and B    |
# | ----- | ----- | ---------- |
# | False | False | False      |
# | False | True  | False      |
# | True  | False | False      |
# | True  | True  | **True**  |
```

```python
# | A     | B     | A or B     |
# | ----- | ----- | ---------- |
# | False | False | False      |
# | False | True  | **True**  |
# | True  | False | **True**  |
# | True  | True  | **True**  |
```

```python
# | A     | not A      |
# | ----- | ---------- |
# | False | **True**  |
# | True  | **False**  |
# # 
```


Operator precedence (highest → lowest):

1. `not`
2. `and`
3. `or`


## 8. Short-Circuit Evaluation

```python
def check():
    print("Function called")
    return True

print(False and check())   # check() is never called
print(True or check())     # check() is never called
```

## 9. Common Logical Mistakes

```python
age = 21

# Wrong:
# if age == 18 or 21:

# Correct:
if age == 18 or age == 21:
    print("Matched")
```

```python
# Wrong:
# if score = 90:

score = 90
if score == 90:
    print("Correct comparison")
```

## 10. Real-World Project: Login System

```python
username = "admin"
password = "python123"

u = input("Username: ")
p = input("Password: ")

if u == username and p == password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

## 11. Real-World Project: BMI Checker

```python
weight = 70
height = 1.75

bmi = weight / (height ** 2)

if bmi < 18.5:
    print("Underweight")
elif bmi < 25:
    print("Normal")
elif bmi < 30:
    print("Overweight")
else:
    print("Obese")
```

## 12. Real-World Project: Discount Calculator

```python
amount = 3200

discount = 20 if amount >= 3000 else 10
final = amount - amount * discount / 100

print("Discount:", discount, "%")
print("Final Amount:", final)
```

## 13. Best Practices


- Keep conditions simple.
- Use descriptive variable names.
- Avoid deeply nested `if` statements.
- Prefer `elif` over multiple independent `if` statements when only one branch should execute.
- Use parentheses to improve readability in complex conditions.
