## 1. Why Do Modules Exist?

Imagine you are building a house. You don't make the bricks and pipes yourself -- you buy them ready-made.

**Modules are the ready-made parts of Python.** Instead of writing everything from scratch, you import code that already exists.

Three reasons modules matter:
1. **Reuse** -- Don't reinvent the wheel. Use tested code.
2. **Organization** -- Split large programs into smaller, manageable files.
3. **Sharing** -- Use libraries written by other developers (thousands free on PyPI).

## 2. What is a Module?

A **module** is simply a `.py` file containing Python code (functions, variables, classes).

A **package** is a folder containing multiple modules + an `__init__.py` file.

The **Standard Library** is the collection of modules that comes built-in with Python.

## 3. Your First Import

The `import` statement loads a module so you can use its code.

````python
import math

print(math.sqrt(16))      # Square root
print(math.pi)            # Pi constant
print(math.factorial(5))  # 5! = 120
````

**Rule:** After `import math`, you access everything with `math.` prefix. This keeps your code organized.

## 4. Different Import Styles

````python
# Style 1: import the module (use module.name)
import random
print(random.randint(1, 6))
````

````python
# Style 2: import specific things (use directly)
from math import sqrt, pi
print(sqrt(25))
print(pi)
````

````python
# Style 3: import with an alias (shorter name)
import math as m
print(m.cos(0))

# Common aliases you will see everywhere:
# import numpy as np
# import pandas as pd
# import matplotlib.pyplot as plt
````

````python
# Style 4: import everything (DANGEROUS -- avoid this!)
from math import *
print(sin(0))  # Works, but you don't know where sin() came from

# Why avoid? It pollutes your namespace -- you might accidentally
# overwrite a function without realizing it.
````

### Which style to use?

| Style | When | Example |
|-------|------|---------|
| `import math` | Most of the time | `math.sqrt(16)` |
| `from math import sqrt` | You need only 1-2 things | `sqrt(16)` |
| `import numpy as np` | The module has a standard alias | `np.array([1,2])` |
| `from math import *` | Almost NEVER | -- |

## 6. Creating Your Own Module

**A module is just a .py file.** Let's create one:

````python
# Step 1: Create a file called 'my_math.py'
module_code = '''
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

PI = 3.14159
'''

with open("my_math.py", "w") as f:
    f.write(module_code)

print("Created my_math.py!")
````

````python
# Step 2: Now we can import and use it!
import my_math

print(my_math.add(5, 3))
print(my_math.subtract(10, 4))
print(my_math.PI)
````

Just like that -- you made your first module! Any `.py` file can be imported.

## 7. The `__name__` Magic

Every Python file has a special variable called `__name__`.

- If the file is **run directly** (`python myfile.py`) -> `__name__` = `"__main__"`
- If the file is **imported** (`import myfile`) -> `__name__` = `"myfile"`

This lets you write code that only runs when the file is executed directly.

````python
# Create a module that demonstrates __name__
code = '''
def greet(name):
    return f"Hello, {name}!"

print("Module loaded. __name__ =", __name__)

if __name__ == "__main__":
    # This code runs ONLY when the file is run directly
    print(greet("Tester"))
'''

with open("greeter.py", "w") as f:
    f.write(code)
print("Created greeter.py")
````

````python
# When we IMPORT greeter, the test code does NOT run
import greeter
print("Imported successfully -- test code was skipped!")
print(greeter.greet("Rahul"))
````

**This is why every real Python file ends with:**
```python
if __name__ == "__main__":
    main()
```

## 8. Packages -- Organizing Multiple Modules

A **package** is a folder with an `__init__.py` file that contains related modules.

````python
# Create a simple package structure
import os
os.makedirs("school", exist_ok=True)

# __init__.py makes 'school' a package
with open("school/__init__.py", "w") as f:
    f.write("# This makes 'school' a package\n")

# Create a module inside the package
with open("school/students.py", "w") as f:
    f.write('''
def get_students():
    return ["Rahul", "Priya", "Amit"]
''')

print("Package 'school' created!")
````

````python
# Now import from the package
from school.students import get_students
print(get_students())

````

This is how real projects are organized:
```
my_project/
    models/
        __init__.py
        user.py
    utils/
        __init__.py
        helpers.py
    main.py
```

## 9. pip -- Installing Third-Party Packages

`pip` is Python's package installer. It downloads packages from [PyPI](https://pypi.org) -- a massive library of free code.

Open your terminal (not this notebook) and type:
```bash
pip install numpy
pip install pandas
pip install matplotlib
```

Once installed, you can import and use them:

````python
# Try importing a third-party package
try:
    import numpy as np
    print("NumPy version:", np.__version__)
except ImportError:
    print("NumPy not installed yet. Run: pip install numpy")
````

### Essential pip commands:

| Command | What it does |
|---------|-------------|
| `pip install pkg` | Install a package |
| `pip uninstall pkg` | Remove a package |
| `pip list` | Show all installed packages |
| `pip freeze > requirements.txt` | Save all packages to a file |
| `pip install -r requirements.txt` | Install from a requirements file |

## 10. How Python Finds Modules

When you `import something`, Python searches these locations in order:
1. **Current directory** (where your script is)
2. **Standard library** (built-in modules)
3. **site-packages** (where pip installs things)

````python
# See where Python looks for modules
import sys
for path in sys.path[:5]:
    print(path)
print("... and more")
````

## 14. Mini Project -- Create Your Own Math Tools Module

````python
# Create a reusable module
# First create the module file
module = open("math_tools.py", "w")
module.write("""# Math utility functions

def is_even(n):
    return n % 2 == 0

def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

def factorial(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

if __name__ == "__main__":
    print("Testing math_tools...")
    print("is_even(10):", is_even(10))
    print("is_prime(17):", is_prime(17))
    print("factorial(5):", factorial(5))
""")
module.close()
print("Module math_tools.py created!")

# Import and use it
import math_tools
print("Is 100 even?", math_tools.is_even(100))
print("Is 29 prime?", math_tools.is_prime(29))
print("7! =", math_tools.factorial(7))
````
