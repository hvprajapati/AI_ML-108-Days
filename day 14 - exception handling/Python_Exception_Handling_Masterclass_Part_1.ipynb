# 1. Why Do We Need Exception Handling?

Imagine you're building an ATM.

- The user enters invalid input.
- The network disconnects.
- The balance file cannot be opened.

Should the program crash?

**No.** We should detect problems and respond gracefully. That's the purpose of exception handling.


# 2. Errors vs Exceptions

**Errors** are problems in your code that prevent execution (such as syntax mistakes).

**Exceptions** are runtime problems that occur while the program is running and can often be handled.

```
Write Code
    │
    ▼
Syntax Correct?
    │
  No ──► SyntaxError
    │
   Yes
    │
    ▼
Program Runs
    │
Runtime Problem?
    │
  Yes ──► Exception
```

## 4. Your First `try` and `except`

````python
try:
    number = int(input("Enter a number: "))
    print(100 / number)
except:
    print("Something went wrong!")
````

Using a bare `except` catches every exception. It's better to catch specific exceptions.

## 5. Catching Specific Exceptions

````python
try:
    number = int(input("Enter a number: "))
    print(100 / number)

except (ValueError,ZeroDivisionError):
    print("Please enter a valid integer")
````

## 6. Catching Multiple Exceptions

````python
try:
    value = int(input("Enter number: "))
    print(10 / value)

except (ValueError, ZeroDivisionError) as error:
    print("Exception:", error)
````

## 7. The `else` Block

````python
try:
    number = int(input("Enter number: "))
except ValueError:
    print("Invalid input")

print("This is a finally block i always exicute")
````

`else` runs only when no exception occurs.

## 8. The `finally` Block

````python
try:
    file = open("sample.txt", "r")
except FileNotFoundError:
    print("File not found")
finally:
    print("Cleanup completed.")
# finally will exicute even after the function returns
````

`finally` always executes, whether an exception occurs or not.

## 10. Nested `try` Blocks

````python
try:
    try:
        print(10 / 0)
    except ZeroDivisionError:
        print("Inner exception handled")
except Exception:
    print("Outer exception handled")
````

## 11. Mini Project - Safe Calculator

````python
try:
    a = float(input("First number: "))
    b = float(input("Second number: "))
    op = input("Operation (+,-,*,/): ")

    if op == "+":
        print(a + b)
    elif op == "-":
        print(a - b)
    elif op == "*":
        print(a * b)
    elif op == "/":
        print(a / b)
    else:
        print("Invalid operation")

except ValueError:
    print("Please enter valid numbers.")
except ZeroDivisionError:
    print("Cannot divide by zero.")
````

## 12. Mini Project - Safe File Reader

````python
filename = input("Enter filename: ")

try:
    with open(filename, "r") as file:
        print(file.read())
except FileNotFoundError:
    print("The file does not exist.")
````

## 13. Common Mistakes

- Using `except:` everywhere.
- Ignoring exception messages.
- Catching exceptions that should be fixed.
- Forgetting `finally` for cleanup.
- Putting too much code inside one `try` block.
