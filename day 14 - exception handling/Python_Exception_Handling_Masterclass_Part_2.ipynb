## 2. The `raise` Statement

````python
age = -5

if age < 0:
    raise ValueError("Age cannot be negative.")
````

`raise` lets you create and throw an exception intentionally when your program detects invalid data.

## 3. Re-raising Exceptions

````python
try:
    x = int("abc")
except ValueError:
    print("Logging error...")
    raise
````

## 4. Creating Custom Exceptions

````python
class InsufficientBalanceError(Exception):
    pass

balance = 1000
withdraw = 1500

if withdraw > balance:
    raise InsufficientBalanceError("Insufficient account balance.")
````

## 5. Exception Hierarchy

## 6. Assertions
assert is simply a way to check if something must be true.

If the condition is True, the program continues.

If the condition is False, Python stops the program and raises an AssertionError.

````python
age = 20

assert age >= 18, "Age must be at least 18."

print("Assertion passed.")
````

Assertions are mainly used during development to catch programming mistakes.

## 7. Exception Chaining (`raise ... from ...`)

````python
try:
    int("hello")
except ValueError as err:
    raise RuntimeError("Conversion failed.") from err
````

Exception chaining preserves the original exception, making debugging easier.

## 8. Logging Exceptions

````python
import logging

logging.basicConfig(level=logging.ERROR)

try:
    10 / 0
except ZeroDivisionError:
    logging.exception("Division failed")
````

## 9. The `warnings` Module

````python
import warnings

warnings.warn(
    "This function will be removed in a future version.",
    DeprecationWarning
)
````

Use **warnings** when execution can continue. Use **exceptions** when execution should stop.

## 10. Best Practices


- Catch specific exceptions instead of using `except:`.
- Keep `try` blocks small.
- Log exceptions in production.
- Use custom exceptions for business rules.
- Always clean up resources (`finally` or `with`).
- Never silently ignore unexpected exceptions.


## 11. Common Anti-Patterns

````python
# Bad Practice
try:
    10 / 0
except:
    pass   # Hides every error!
````

````python
# Better
try:
    10 / 0
except ZeroDivisionError as error:
    print(error)
````

## 12. Mini Project - Banking System

````python
class InsufficientBalanceError(Exception):
    pass

balance = 5000

def withdraw(amount):
    global balance

    if amount <= 0:
        raise ValueError("Amount must be positive.")

    if amount > balance:
        raise InsufficientBalanceError("Not enough balance.")

    balance -= amount
    return balance

try:
    print(withdraw(1000))
except Exception as e:
    print(e)
````

## 13. Mini Project - Student Marks Validator

````python
def validate_marks(mark):
    if not 0 <= mark <= 100:
        raise ValueError("Marks must be between 0 and 100.")
    return mark

try:
    print(validate_marks(105))
except ValueError as e:
    print(e)
````

## 14. Mini Project - Login Validation

````python
class InvalidCredentials(Exception):
    pass

username = "admin"
password = "python123"

u = input("Username: ")
p = input("Password: ")

try:
    if u != username or p != password:
        raise InvalidCredentials("Invalid username or password.")
    print("Login successful!")
except InvalidCredentials as e:
    print(e)
````
