# Introduction to Tuples

## What is a Tuple?
A tuple is an ordered and immutable collection of elements.

```python
t=(10,20,30)
print(t)
```

# Creating Tuples

## Creating Tuples
Different ways to create tuples.

```python
t1=()
t2=tuple()
t3=(1,2,3)
t4=('Python',22,True)
print(t1,t2,t3,t4)
```

# Single Element Tuple

## Single Element Tuple
The comma creates the tuple.

```python
t=(10,)
print(type(t))
```

# Tuple Packing

## Tuple Packing
Python automatically packs values into tuples.

```python
t=10,20,30
print(t)
```

# Tuple Unpacking

## Tuple Unpacking

```python
t=(10,20,30)
a,b,c=t
print(a,b,c)
```

# Extended Unpacking

## Extended Unpacking

```python
t=(10,20,30,40,50)
a,*mid,b=t
print(a)
print(mid)
print(b)
```

# Indexing

## Indexing
Positive and negative indexing.

```python
t=(100,200,300,400)
print(t[0])
print(t[-1])
```

# Slicing

## Slicing

```python
t=(10,20,30,40,50)
print(t[1:4])
print(t[::-1])
```

# Immutability

## Tuples are Immutable

```python
t=(10,20,30)
# t[1]=999
```

# Traversing Tuples

## Traversing Tuples

```python
t=(10,20,30)
for item in t:
    print(item)
```

# Concatenation

## Tuple Concatenation

```python
a=(1,2)
b=(3,4)
print(a+b)
```

# Repetition

## Tuple Repetition

```python
print((1,2)*3)
```

# Membership

## Membership Operators

```python
print(20 in (10,20,30))
```

# count() Method

## count()

```python
t=(1,1,1,2,2,3)
print(t.count(1))
```

# index() Method

## index()

```python
t=(10,20,30)
print(t.index(20))
```

# Built-in Functions

## Built-in Functions

```python
t=(10,20,30,40)
print(len(t))
print(max(t))
print(min(t))
print(sum(t))
```

# Nested Tuples

## Nested Tuples

```python
t=((1,2),(3,4),(5,6))
print(t[1][0])
```

# Frequency Counter

## Frequency Counter

```python
t=(1,1,2,2,2,3)
freq={}
for item in t:
    freq[item]=freq.get(item,0)+1
print(freq)
```
